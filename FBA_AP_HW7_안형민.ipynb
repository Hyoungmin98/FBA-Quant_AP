## Chapter 13. Empirical Evidence on Security Returns  

### 13.1 Two-Pass Tests of Asset Pricing  

- 자산 가격결정 모형이 자산의 위험과 수익률 간의 관계를 설명하는 데 얼마나 유효한지 평가하기 위한 방법  

1. First Pass Regression  
- 각 자산의 초과 수익률을 시장 요인이나 위험 요인에 회귀 분석  
$R_i - R_f = \alpha_i + \beta_i (R_m - R_f) + \epsilon_i$  

2. Second Pass Regression  
- 각 자산의 평균 초과 수익률을 첫 번째 단계에서 계산된 베타 값에 대해 회귀 분석  
$\bar{R}_i - R_f = \gamma_0 + \gamma_1 \beta_i + u_i$  

- 주식 수익률은 극도로 변동성이 높기 때문에 평균 수익률 테스트의 정밀도가 떨어진다는 단점 존재  

- 주식에만 배타적으로 초점을 맞추면 부동산, 민간 기업, 상품 또는 인적 자본과 같이 총체적으로 주식 시장보다 훨씬 더 큰 가치를 지닌 중요한 자산군이 제외된다.  

- 투자자는 CAPM의 단순 버전에서 가정하는 것처럼 무위험 이자율로 차입할 수 없다.

### 13.5 The Equity Premium Puzzle   

중요한 것은 부(재산) 자체보다는, 그 부가 얼마나 지속적으로 소비를 지원할 수 있느냐이다.  
→ CAPM에서 위험을 측정할 때 시장 수익률 대신 소비 흐름과 자산 수익률 간의 관계를 고려하는 것이 더 바람직할 수 있다.  

$E(r_M) − r_f = ACov(r_M, r_C)$  
C: 소비 추적 포트폴리오  
A: 위험 회피 성향  

- 문제는 소비 증가율이 그리 가변적이지 않다는 점이다. 소비 변동성이 낮기 때문에 공분산 항도 낮게 나온다.   
- 이처럼 "체계적" 소비 리스크가 낮기 때문에, 아주 높은 수준의 위험 회피를 가정하지 않는 한, 역사적으로 관찰해 온 것만큼 큰 시장 리스크 프리미엄을 정당화하기는 어렵다.  
→ 이러한 경험적 문제를 Equity Premium Puzzle이라고 한다.  

그럼에도 불구하고 Jagannathan과 Wang은 소비 추적 포트폴리오에서 어느 정도 성공을 거두었다.  

  


※ Expected versus Realized Returns  

Fama와 French는 주식 시장에서 실제로 실현된 수익률(average realized returns)을 기반으로 리스크 프리미엄을 추정하는 것이 문제가 될 수 있다고 생각했음.  

배당금 할인 모델(DDM)을 활용하여 기대 수익률을 추정.  
- 1872-1949년 기간 동안 배당금 할인 모델(DDM)이 평균 실현 초과 수익률과 유사한 예상 위험 프리미엄의 추정치를 산출한다는 것을 발견  
- 1950-1999년 기간 동안, DDM은 훨씬 더 낮은 기대 위험 프리미엄을 산출했는데 이는 당시의 실현된 수익률이 매우 높았으며, 투자자들이 실제로 기대했던 수익률보다 더 큰 수익을 거두었다는 것을 의미한다.  

  
$D_1$: 연말 배당금  
$P_0$: 주식의 현재 가격  
$g$: 예상 배당 성장률  

※ Survivorship Bias  
평균적인 미국 데이터를 사용하면 기대 수익률 추정치에 일종의 생존 편향이 부여될 수 있다는 것인데, 이는 다른 많은 국가와 달리 미국은 이러한 극단적인 문제의 희생자가 된 적이 없기 때문이다.    

평균 수익률이 가장 높은 국가의 경험을 바탕으로 리스크 프리미엄을 추정하고 전체 표본 기간 동안 살아남지 못한 주식 시장의 증거를 무시하면 기대 수익률 추정치에 상향 편향이 발생할 수 있다.  


※ Behavioral Explanations of the Equity Premium Puzzle  
Barberis와 Huang은 비합리적인 투자자 행동이 주식 위험 프리미엄 퍼즐과 주식 시장 참여 퍼즐을 설명할 수 있다고 주장하며, 투자자들이 손실을 두려워하고, 개별 위험을 과대평가하기 때문에 더 높은 위험 프리미엄을 요구한다고 설명한다.  

**loss aversion**: 투자자들이 손실을 이익보다 더 크게 받아들이는 경향이 있다.    
**narrow framing**: 투자자들이 개별 위험을 전체 포트폴리오의 맥락이 아닌 독립적으로 평가하는 경향이 있다.  
→ 이러한 이유로 투자자들은 합리적인 모델이 예측하는 것보다 더 높은 위험 프리미엄을 요구한다.

※ Style Analysis  

- 윌리엄 샤프(William F. Sharpe)가 개발한 방법으로, 펀드 매니저가 운용하는 포트폴리오의 성과를 분석하고, 그 포트폴리오의 자산 배분이나 투자 스타일을 알아내기 위한 방법.  

포트폴리오의 수익률을 여러 자산 클래스의 수익률을 사용하여 회귀 방정식을 만들고 각 자산 클래스의 회귀 계수는 해당 자산에 대한 가중치, intercept(alpha)는 매니저의 능동적 투자 성과를 나타낸다.   

회귀 방정식이 아닌 최적화 기법을 사용하는 경우, 회귀 모델 대신 포트폴리오와 벤치마크 간의 추적 오차(Tracking Error)를 최소화하는 가중치를 찾는 방식이 사용될 수 있다. 이때 회귀 방정식과는 달리 각 자산의 가중치가 모두 양수이거나 자산의 가중치의 합이 1이라는 제약 조건을 설정할 수 있다.  



※ Cap-weighted benchmarks  
- 각 종목의 시가총액을 기준으로 가중치를 부여하는 방식  

시가총액 가중 지수가 지수로는 완벽하지만 벤치마크 포트폴리오로는 다소 비효율적이다.  
→ efficient frontier를 위험 단위당 가장 높은 보상을 제공하는 효율적인 포트폴리오를 찾는 곳이라고 한다면 시가총액 가중 지수는 efficient frontier 내부에 있을 것이다.  


동일 가중 포트폴리오가 시가총액 가중 포트폴리오에 비해 동일 수준 리스크에 대해 더 나은 성과를 보인다.  
→ 시가총액 가중 지수는 시가총액이 큰 종목을 과대평가 하기 때문에 다각화가 잘 이루어지지 않는다.  


Cap-weighted benchmarks  단점 정리  
1. 시가총액이 큰 종목을 과대평가 하기 때문에 다각화가 잘 이루어지지 않는다.  
2. 최적이 아닌 비효율적인 암묵적 요인 노출을 보이는 경향이 있다.  

※ Smart-Weighted benchmarks  
전통적인 시가총액 가중 방식의 단점을 보완하기 위해 등장한 전략으로, 다양한 투자 요인(가치, 모멘텀, 저변동성 등)을 반영하여 자산 배분을 최적화하려는 방식.  

1. 포트폴리오에 포함시키고 싶은 요인 노출도를 선택(ex. 모멘텀, 거래량, 가치, 변동성) 
2. 다각화 전략 선택(ex. 동일 가중치, 리스크 패러티)   

결과적으로 Smart-Weighted benchmark가 Cap-weighted benchmark보다 더 높은 수익률과 샤프비율, 낮은 변동성을 가진다.
