# Chapter 2. The Fundamentals of QEPM   
## 2.1 INTRODUCTION  
퀀트 주식 포트폴리오 매니저는 일반적으로 벤치마크를 능가하는 성과를 목표로 한다.  
→ Alpha 추구하기 위해 QEPM의 7가지 원칙을 따른다.  
- 금융 시장이 완전히 효율적으로 운영되지 않는다.  
- QEPM의 통계 집약적 방법과 직접적인 관련이 있는 통계적 문제를 고려한다.   

## 2.2 QEPM Alpha  
Alpha: 기준 상품의 수익률에 대한 포트폴리오의 초과수익률   
→ 알파를 높인다는 것은 참조 상품에 대한 직접적인 노출 위험을 증가시키지 않으면서 포트폴리오의 수익률을 높이는 것  

참조 상품이 포트폴리오 매니저의 벤치마크인 경우: $\alpha^B$    

통계적 기법을 사용하면 포트폴리오 수익률을 Expected Return 와 Residual Return 으로 분류할 수 있다.  

1. Benchmark α  
  
→ alpha + epsilon이 Residual Return, $\beta r_B$가 Expected Return  

퀀트 포트폴리오 매니저의 목표는 위험조정 수익률을 높이는 것이기 때문에 Residual Return(잔여수익률)이 중요하다.  

벤치마크 α은 잔여 수익률의 기대값이고, 잔여 수익률의 두 번째 구성 요소인 ε는 잔여 수익률과 평균의 편차이다.  

2. CAPM α  
  
이 방정식에서는 α를 CAPM α라고 부른다.   
- CAPM 이론에 따르면 CAPM α는 0과 같아야 한다. 유의한 양수값을 가지는 경우 포트폴리오 매니저는 초과 위험 조정 수익률을 제공하고 있는 것.  

3. Multifactor α  
   
이α는 주식 수익률에 영향을 미치는 많은 요소가 있는 모델을 사용하여 만들어진다.  

매니저가 비효율적인 벤치마크를 사용한다고 가정해 보자면, 이는 시장 수익률에 대한 벤치마크의 α(벤치마크 자체의 알파)가 음수가 된다.  
결과적으로 포트폴리오의 벤치마크 α는 포트폴리오의 시장α(CAPM α)보다 높을 것.  
따라서 비효율적인 벤치마크를 사용하면 포트폴리오의 성과가 전체 시장에 대해 측정된 경우보다 더 좋아보이게 된다.  


- 긍정적 벤치마크 α이 차익거래 가격 이론(APT)과 일치한다.   
  
→ 주식 수익률의 실제 요인을 알고 있고 f1이 벤치마크(ex. S&P500)이라고 가정.  
또한 APT 모델의 다른 모든 요인이 첫번째 요인(벤치마크)과 상관관계가 없는 경우 벤치마크 alpha는 다음과 같다.  
  
벤치마크 alpha에는 다른 모든 요인의 효과가 포함되기 때문에 양수일 가능성이 크다.  

만약 일반적으로 벤치마크가 다른 모든 요인과 상관관계가 있을 경우 벤치마크 alpha는 다음과 같다.  
  
벤치마크 α에는 벤치마크를 제외한 모든 요인의 효과가 포함되므로 벤치마크 α가 양수일 가능성이 매우 높다.  

따라서 긍정적인 벤치마크 α은 관리자가 통계적 차익 거래에 대해 보상을 받는 APT 모델과 일치한다.  


α 그 자체로 포트폴리오의 초과 수익률의 주요 척도이지만, 포트폴리오의 잔여 위험으로 α를 조정하는 핵심 지표인 정보 비율도 활용할 수 있다.  
  
$\alpha^B$는 포트폴리오 매니저의 예상되는 알파이며, $w$는 예상되는 추적 오차  


2.2.5 Ex-Ante and Ex-Post α

- 어떤 유형의 α를 사용하든, 우리가 미래에 달성하기를 기대하는 값과 실제로 실현되는 값 사이에는 차이가 존재  
- 사전적 α는 예상되는 α이고, 사후적 α는 실제로 이루어진 α.  
- 사전적 α는 정량적 포트폴리오 매니저가 자신의 포트폴리오를 구성할 때 관심의 대상  

## 2.3 THE SEVEN TENETS OF QEPM  
퀀트 주식 포트폴리오 매니저는 높은 알파와 높은 정보비율이라는 두가지 목표를 두고 있다.  
이를 달성하기 위한 QEPM의 7가지 원칙은 다음과 같다.  
1. 시장은 대부분 효율적이다.  
2. 순수 차익거래 기회는 존재하지 않는다.  
3. 정량적 분석은 통계적 차익거래 기회를 창출한다.  
4. 정량 분석은 사용 가능한 모든 정보를 효율적인 방식으로 결합한다.  
5. 정량적 모델은 건전한 경제 이론에 기초해야 한다.  
6. 지속적이고 안정적인 패턴을 반영해야 한다.  
7. 벤치마크에서 포트폴리오의 편차는 불확실성이 충분히 작은 경우에만 정당화된다.  

## 2.4 TENETS 1 AND 2: MARKET EFFICIENCY AND QEPM  
시장이 완전 효율적이라면 QEPM은 무의미한 행위에 그치기 때문에 액티브 포트폴리오 매니저는 시장에 자신이 이용할 수 있는 비효율성이 존재한다고 믿어야 한다.  

- 1970년대, 시카고 대학의 교수인 유진 파마는 시장 효율성에 대한 검증 가능한 기준을 설정  → 약형, 준강형, 강형의 세 가지 효율성 수준을 정의했으며, 이는 시장 가격이 정보의 점점 더 많은 부분을 반영한다는 것을 의미  

약형 효율성 (Weak form): 증권 가격이 이전의 증권 가격에서 사용 가능한 정보를 반영   
준강형 효율성 (Semistrong form): 증권 가격이 공개적으로 이용 가능한 모든 정보를 반영  강형 효율성 (Strong form): 증권 가격이 공개적이든 사적으로 보유하고 있든 모든 정보를 반영  


- 약형 효율성에서는 과거 주식 가격이 투자자에게 실행가능한 정보를 제공하지 않기 때문에 기술적 분석이 작동하지 않는다.  
- 모멘텀의 패턴이 유지되는 경우 시장의 약형 효율성은 존재하지 않는다.  

- 준강형 시장 효율성에서는 퀀트 포트폴리오 매니저와 다른 투자자들이 어떤 공개 정보를 사용해도 위험 조정 초과 수익률을 얻을 수 없다는 것을 의미한다.(ex. Value)  

- 강형 효율성은 검증하기 가장 어려운 효율성 형태로 사적 정보조차도 투자자에게 일관된 우위를 제공하지 않는다.  
- 법에서 내부 정보를 기반으로 거래하는 행위를 금지하는 것이 효과적이라면 시장은 강형 효율적일 수 없다.  

일부 액티브 매니저들이 지속적인 초과수익을 내는 것은 시장이 완전히 효율적이지 않다는 것을 확신시킨다.  

※ 약형 이상 현상  
- 약형 시장 효율성을 검증하는 실험들은 과거 가격을 사용하여 개별 주식의 미래가격을 예측할 수 있는지 테스트한다.  
- 투자자들은 모멘텀을 이용하여 이전 기간에 수익률이 높은 주식을 매수함으로써 이익을 얻을 수 있음  
→ 한 연구에 따르면 1962년 7월~1994년 12월까지 특정 주가지수의 자기상관성은 약 35%였다. 일주일 또는 한달의 기간에서는 월별 주식 수익률이 음의 자기상관성을 보이기도 하였으며 산업 수익률이 양의 자기상관성을 보일 수 있다는 증거도 존재한다.  

- 트레이더들은 이러한 것들이 효과가 있다고 믿지만, 기술적 지표의 테스트 결과는 이러한 믿음을 명확히 확인하지 못한다.  

※ 준강형 이상현상  
- 시장이 준강형 효율적이지 않다는 증거를 주는 많은 이상현상들이 존재한다.  
- 대표적으로 '어닝 서프라이즈'로, 예상보다 높은 수익을 보고하는 주식들은 발표 후 몇 주 동안 위험 조정된 초과 수익을  보이는 경향이 있다.  
→ 연구에 따르면 양의 어닝 서프라이즈 후 13~26주 동안 초과 수익을 보였다.  
- 또다른 이상현상으로 1월 효과 존재. 소형주 또는 전년도 성과가 좋지 않았던 주식은 1월 첫 주에 수익률이 매우 높은 경향이 있다.  
→ 1. 기관 및 개인 투자자들이 자본 손실을 발생시켜 자본 이득을 상쇄하고 소득세를 상쇄하기 위해 12월에 성과가 좋지 않은 주식을 매도하는 tax loss harvesting 때문이다. 이후 투자자들은 새해가 되면 해당 주식을 다시 매수한다.  
2. 펀드매니저들은 증권거래위원회에 그들의 보유 주식을 보고하기 전 연말에 손실 주식을 매도하는 윈도우 드레싱 때문   
3. 펀드매니저들은 해의 초반에 베팅을 통해 대부분의 기금에서 주는 연간 보너스를 받기 위한 좋은 출발을 목표로 한다.  

※ 1월 효과는 차익거래자들과 퀀트 매니저들이 12월에 미리 거래를 시도함에 따라 12월로 효과가 옮겨가고 있다.  

- 이외에도 월요일에 시장이 약간 하락하는 경향을 가지는 월요일 효과가 존재한다.  

※ 펀더멘털 데이터 관련 이상현상  
- 회사의 재무제표에서 얻은 데이터  
- 낮은 PBR 또는 낮은 PER을 가진 주식 포트폴리오가 높은 위험 조정 수익률을 기록한다.  

최근 몇 년 동안 새로운 이상 현상들이 보이며, 일부 오래된 이상 현상들이 다시 유행하고 있다.  
- 저변동성, 저베타 이상 현상이 다시 인기를 얻고 있다.  
- 최근에는 유동성 및 혼잡과 관련된 팩터들이 부각되고 있다.  
→ 혼잡은 여러 시장 참가자들이 동일 거래를 집중적으로 따르기 시작하여 유동성이 취약해지고 거래의 위험 및 수익 역학이 변화하는 투자에서 최근 발견된 위험. 덜 혼잡한 주식을 매수하는 전략이 초과 수익률을 기록한다.  

※ 강한 이상현상  
- 강형 시장 효율성 테스트 중 하나로 '내부자 거래 테스트' 가 있다.  
- 임원들은 자신들이 소유한 회사의 증권 매수 및 매도를 SEC에 보고해야 한다.  
- 외부인이 내부자가 주식을 살 때 매수하고, 내부자가 팔 때 매도했다면, 위험 조정된 초과 수익을 얻을 수 있다.  

비효율적인 시장을 완전히 이해하기 위해서는 CAPM에서 멀티팩터 모델로 전환하는 것이 필요할 수 있다.  

행동 재무론은 심리학과 비합리적 행동이 투자 결정에 미치는 영향을 연구함으로써 이상 현상을 설명하려고 시도한다.  
→ 모호성 회피, 가용성 편향, 확증 편향, 처분 효과, 물타기 편향, 지식의 착각 등   


2.4.3 Market Efficiency and QEPM  
시장이 완벽하게 효율적이지 않고, 정량적 주식 포트폴리오와 같은 액티브 포트폴리오 운용이 가치가 있는 이유는 다음과 같다.  

1. 정보를 얻는 것은 비용이 들지만 모든 사람이 정보를 얻기 위해 비용을 지불할 수 있거나 그렇게 하려는 의지가 있는 것은 아님  
2. 공개 정보조차도 시장을 통해 다소 느리게 전달됨   
3. 모든 투자자가 대량의 정보, 특히 정량적 정보를 처리할 능력이 있는 것은 아님  
4. 공개 정보를 필터링함으로써 일부 사람들은 사실상 사적인 정보를 만들어냄  
5. 일부 투자자들은 정보의 논리적 해석보다는 감정에 기반하여 투자 결정  
6. 타인의 추정된 비합리성을 이용하려는 일부 시도는 실제로 더 많은 비효율성을 만들어냄  
7. 경제 상황, 특히 기술의 상태는 항상 변하며, 사람들이 이러한 변화에 적응하는 데 시간이 걸림  
8. 거래 비용은 경제 모델과 현실 사이의 격차를 만듦  
9. 세금은 시장에서 왜곡을 일으킴  
10. 금융 시장에 대한 정부 규제는 경제 모델과 현실 사이의 격차를 만듦    


정보가 공개적으로 접근 가능해지면, 퀀트 포트폴리오 매니저들은 그 정보를 활용하는 데 가장 좋은 위치에 있다.  
→ 퀀트 매니저들은 전자 데이터 서비스를 통해 즉시 뉴스 알림을 받는 반면 평균적인 투자자들은 TV나 신문 헤드라인을 통해 알게 된다.    
→ 퀀트 매니저들은 대량의 데이터 영역을 비교적 빠르게 분류하는 스킬을 가지고 있기에 이를 반영하여 수익률 모델을 업데이트 할 수 있다.  

이것들은 시장이 완벽하게 효율적이지 않은 필수적인 이유들로, 퀀트 주식 포트폴리오 매니저들에게 이익이 된다.  
→ 퀀트 포트폴리오 매니저들과 다른 차익거래자들로 인해 시장의 비효율성을 제거하지 않은 이유는?  
1. 비효율성이 순수한 차익거래 기회가 아니라 제로가 아닌 저위험 차익거래 기회였을 가능성  
2. 충분한 투자력을 가진 차익거래자가 시장에 충분하지 않았을 가능성   

## 2.5 TENETS 3 AND 4: THE FUNDAMENTAL LAW, THE INFORMATION CRITERION, AND QEPM  
- 액티브 관리의 기본 법칙  
$IR = IC \cdot \sqrt{BR}$  
정보비율(IR)은 정보계수(IC)와 범위(BR)의 제곱근을 곱한 값이다.  

$\frac{\alpha^B}{\omega} = IC \cdot \sqrt{BR}$  
정보 비율을 높이기 위해서는 정보계수(IC)를 증가시키거나 범위(BR)를 확장해야 한다.  
→ 퀀트 매니저들에게는 이미 모델에 있는 팩터보다 더 중요한 팩터를 찾음으로써 IC를 증가시킬 수 있고 모델에 이미 있는 것과 상대적으로 상관관계가 적은 더 많은 팩터를 찾음으로써 BR을 증가시킬 수 있다.  


2.5.1 The Truth about the Fundamental Law  
- QEPM의 핵심 작업중 하나는 수익률과 팩터 간의 관계를 명시하는 모델을 추정함으로써 미래 주식 수익률을 예측하는 것이다.  
$r_{it} = \alpha_i + \beta_{i1}f_{1t} + \beta_{ik}f_{jt}+\epsilon_{it}$  

1. 정보비율의 제곱은 수익률 예측 방정식의 적합도($R^2$)과 비슷하다.  
2. 범위(BR)는 해당 회귀식에 사용된 독립변수의 개수  
3. 정보계수의 제곱은 $R^2$을 증가시키는 데 각 독립변수가 평균적으로 기여하는 정도  
4. 기본 법칙(fundamental law)은 모델의 적합도를 독립변수의 수와 각 변수의 평균 기여도로 분해  
5. 벤치마크를 무시하고 포트폴리오 수익률에서 무위험 이자율을 차감할 경우 정보비율은 사실상 달성 가능한 최대 샤프비율이 된다. 기본 법칙은 최대 샤프비율을 독립변수의 수와 각 변수의 평균 기여도로 분해.  

2.5.2 The Information Criterion  
- 기본 법칙에 대한 하나의 오해는 이 법칙이 모든 액티브 관리에 적용된다는 것이다.  
→ 기본 법칙은 포트폴리오 매니저가 최적의 포트폴리오를 생성할 때만 적용. 하지만 많은 포트폴리오 매니저들은 모든 가능한 정보를 가장 효율적인 방법으로 사용하지 않기 때문에 최적이 아닌 포트폴리오를 생성한다. 이는 원칙4를 위배하게 된다.  

- 포트폴리오 매니저가 애널리스트-리비전 전략을 따를 수 있다.  
→ 지난달에 애널리스트 평가가 개선된 주식들로 구성된 동일가중 포트폴리오 구축  
이때 포트폴리오의 포함되는 것(더미변수)을 변수로 고려하여 다음의 방정식을 추정한다.  
$r_{i, t+1}=\alpha + \beta_{it}f+\epsilon_{i,t+1}$   
f값이 양수라면 이 회귀식은 평가가 개선된 주식들이 미래에 더 높은 수익률을 보일 것을 나타낸다.  
하지만 이 경우에 포트폴리오 매니저는 모든 가능한 정보를 사용하지 않고 있다.  

위의 식에서 $\beta_{it} = 1$인 주식들이 더 높은 기대 수익률을 가질 것이라고 말하는 것뿐만 아니라, 어떤 주식들이 더 높은 위험(변동성)을 가지는지도 식별할 수 있다.  
→ 이 정보를 사용하지 않았기 때문에 정보 기준을 충족하지 않는다.  

2.5.3 Information Loss  
포트폴리오 매니저가 사용 가능한 모든 정보를 사용하여 달성할 수 있었던 정보비율과 실제로 사용 가능한 정보의 일부만을 사용하여 얻은 정보비율을 비교함으로써 매니저가 놓친 정보의 양을 정량화할 수 있다. → 정보손실(IL)  
$IL = 최대\ IR - 실제\ IR$   
ex. 정보 손실이 0.1이라면, 포트폴리오 매니저는 자신이 감수하는 위험의 10%에 대해 1%의 잠재적인 추가 수익을 놓치고 있다는 것을 의미한다.  

## 2.6 TENETS 5, 6, AND 7: STATISTICAL ISSUES IN QEPM  
QEPM의 원칙 5, 6, 7은 포트폴리오 선택 문제에 통계 기법을 적용함으로써 발생하는 문제들에 관한 것이다.  
→ 이 섹션에서는 포트폴리오 매니저가 무시할 수 없는 세가지 주요 문제인 데이터 마이닝, 변수 안정성, 변수의 불확실성에 대해 논의   

2.6.1 Data Mining  
- 데이터 마이닝은 퀀트 모델이 탄탄한 경제 이론에 기반해야 한다고 요구하는 원칙 5를 위반한다.  
- 데이터 마이닝은 역사적 주식 수익률에 대해 많은 팩터 조합에 대해 회귀분석을 수행할 때 나타난다.  
ex. 회귀분석 과정에서 독립변수의 양을 너무 많이 추가하여 $R^2$값이 100%가 되는 현상  

- 문제가 명백하지 않은 경우도 존재한다.  
→ 단계별 회귀를 수행한다고 가정하고 100개 변수 중 가장 유의미한 변수를 의도적으로 찾는다고 하면 선택 가능한 변수 풀이 매우 크기 때문에 반드시 유의미한 변수를 얻을 수 있다. 따라서 변수의 통계적 유의성이 실제 주식 수익률을 잘 설명한다는 실질적 징표라고 할 수는 없다.  

- 개별 연구자가 의도적으로 변수를 선택하지 않더라도, 연구자 커뮤니티 전체가 집단적으로 단계별 회귀를 수행할 수 있다.  

따라서 특정 팩터를 테스트할 때 이전에 사용되지 않은 데이터를 사용하지 않는 한 데이터마이닝이나 데이터 스누핑을 완전히 피할 수는 없다.  

※ 퀀트 모델은 사용한 팩터들이 주식 수익률을 잘 설명하고, 수익률과의 관계가 이미 다른 투자자들에 의해 이용되지 않았다고 생각할 충분한 이유가 있을 때만 작동할 것   


2.6.2 Parameter Stability  
- 과거 데이터에 기반하여 예측을 하기 위해서는 역사가 반복된다고 가정해야 함  
→ 하지만 시장은 지속적으로 변화하기 때문에 역사적 패턴이 유지되지 않는다.  

- 지난 십 년 동안, 팩터 수익률 안정성에 대한 위협이 증가했으며, 정량적 투자와 관련된 위험도 증가했다.  
→ 이는 부분적으로 '군중화'현상 때문  
군중화는 여러 시장 참여자들이 같은 거래를 따르면서 유동성이 취약해지고 거래의 위험 및 수익률 역학이 변경될 때 발생한다.  

- 미래 전략의 성과를 평가하기 위해 포트폴리오 매니저는 역사적 백테스팅이 현실적인지, 군중화를 인식하는지, 팩터 자체의 본질이 시간이 지남에 따라 어떻게 변하는지 인식해야 한다.   

2.6.3 Parameter Uncertainty  
액티브 포트폴리오가 겉으로 보기에 위험하지 않더라도, 높은 양의 불확실성을 가질 수 있으며, 이는 위험의 구성 요소이다.  

표준오차(모델에서의 오차의 분산과 설명변수의 분산)를 고려하면, 우월해 보이는 액티브 포트폴리오의 위험 조정 수익률이 실제로는 벤치마크의 것과 다를 바 없음이 명백해질 수 있다.   

- 일반적으로 포트폴리오 매니저는 추정치의 값뿐만 아니라 추정치의 표준오차도 고려해야 한다.

- QEPM은 시장이 일반적으로 효율적으로 기능하지만 많은 비효율성의 패턴을 포함하고 있으며 이는 통계적 차익거래의 기회를 제공한다는 믿음에 기반한다.  
- 기본 법칙에서 정보 기준과 정보 손실 개념에 대한 이해는 포트폴리오 매니저들이 이러한 기회를 최대한 활용하는 데 도움을 준다.

## 3.1 INTRODUCTION

## 3.2 BASIC QEPM MODELS AND PORTFOLIO CONSTRUCTION PROCEDURES  

현대 금융 경제학의 핵심 아이디어: 주식의 평균수익률이 위험을 감수하는 대가로 작용한다.  
- 팩터 모델에서 평균 주식수익률은 주식이 그 팩터가 대표하는 위험에 대한 노출과 그 위험에 대한 노출 당 보상에 비례한다는 것을 보여준다.  

**펀더멘털 팩터 모델**: P/E ratio와 시가총액 등의 펀더멘털 팩터 활용  
**경제적 팩터 모델**: GDP, 인플레이션 등의 거시경제 변수를 위해 개발되었지만 다른 유형의 팩터들도 다룰 수 있음.   

- 펀더멘털 팩터 모델의 경우 팩터 익스포져를 알고 있는 상태에서 팩터 수익률을 추정, 경제적 팩터 모델의 경우에는 팩터 수익률을 기반으로 팩터 익스포져 추정.  

펀더멘털 팩터 모델에서 팩터 익스포져는 회사의 재무제표 및 기타 데이터 소스에서 직접 읽을 수 있다.   

### 3.2.1 Factor Choice  
- 펀더멘털 팩터 모델은 일반적으로 펀더멘털 팩터들만을 고려하기 때문에 팩터 선택을 제한한다.  
- 경제적 팩터 모델은 더 유연하며 경제 팩터와 모든 펀더멘털 팩터를 허용한다.  

### 3.2.2 The Data Decision  
위 단계에서 K개의 팩터를 선택했다고 가정.  

데이터 세트에는 횡단면 차원과 시계열 차원이라는 두 가지 차원이 존재한다.  
- 횡단면 차원은 데이터 세트를 구성하는 주식의 특성으로 정의. 횡단면 차원의 데이터는 최종 포트폴리오 특성에 영향을 미치고 데이터 세트에 포함된 주식 수는 추정 용이성에 영향을 준다.  
- 시계열 차원은 데이터의 주기성과 전체 데이터 세트의 기간에 관한 것.

모델과 데이터 세트를 고려할 때 포트폴리오 매니저는 몇가지 결정을 더해야 한다.  
- 벤치마크 대비 성과를 측정하는 경우가 일반적이므로, 모델에서 벤치마크 자체를 팩터로 포함할지 여부  
- 총 수익률을 사용할지 아니면 로그 수익률을 사용할지, 무위험 이자율을 초과하는 수익률을 사용할지 결정    

### 3.2.3 Factor Exposure  
매니저가 팩터를 선택하고 데이터 세트도 정의했다면 팩터 노출과 팩터 프리미엄을 결정할 수 있다.  

펀더멘털 팩터 모델: ex. 팩터가 P/B ratio라면 주식 i의 팩터노출은 주식 i의 P/B ratio의 최근 관찰값  

경제적 팩터 모델: 팩터 프리미엄을 구하고 데이터 기간의 모든 시간 간격에서 주식 i의 수익률을 해당 팩터 프리미엄에 대해 회귀분석  

### 3.2.4 Factor Premium  
펀더멘털 및 경제적 팩터 모델을 나타내는 방정식은 $r_{i} = \alpha_i + \beta_{i1}f_{1} + \beta_{ik}f_{K}+\epsilon_{i}$로 동일하다.  

펀더멘털 팩터 모델의 팩터 프리미엄은 단면 회귀분석 또는 패널 회귀분석을 통해 추정  

경제적 팩터 모델의 팩터 프리미엄은 팩터의 성격에 따라 다양한 방식으로 결정된다.  
→ 거시경제 팩터의 경우, 변수 자체의 값이 팩터 프리미엄으로 간주. 이 값이 정확한 프리미엄은 아니지만 정확한 프리미엄과 비례하기 때문에 이 비율 자체를 팩터 프리미엄으로 대체한다.  

+ 더 일반적으로 팩터 프리미엄은 제로 투자 포트폴리오를 구성함으로써 계산된다.  
ex. 모델의 팩터가 사이즈라고 하면 사이즈 값을 정렬하여 상하위 롱숏한 포트폴리오의 수익률을 팩터 프리미엄으로 간주.  

### 3.2.5 Expected Return  
추정값은 본질적으로 우리가 주식의 팩터 노출과 팩터 프리미엄을 기반으로 주식의 예상 수익률을 결정할 수 있게 한다.  

### 3.2.6 Risk  
모델의 추정은 또한 위험에 대한 중요 정보를 제공한다.  
→ 체계적 위험 & 비체계적 위험  
체계적 위험. 즉 분산할 수 없는 위험은 투자자들에게 주된 관심사로 $\alpha_i + \beta_{i1}f_{1} + \beta_{ik}f_{K}$로 표현된다.  
비체계적 위험, 분산할 수 있는 위험은 $\epsilon_{i}$에 의해 포착된다.  

### 3.2.7 Forecasting  
모델에서의 팩터 프리미엄과 팩터 노출은 과거 데이터에서 결정되기 때문에 수익률, 노출, 프리미엄 간의 관계는 미래에 변화할 가능성이 높다.  
→ 따라서 일반적으로 팩터 프리미엄을 예측하는 것이 필요(?)  

## 3.3 THE EQUIVALENCE OF THE BASIC MODELS    
펀더멘털 팩터 모델과 경제적 팩터 모델은 주식의 기대 수익률이 펀더멘털 팩터 노출의 선형 함수일 경우 서로 동등하다.  
→ 팩터 모델의 동등성  

따라서 펀더멘털 팩터 모델이 올바른 모델이라면 펀더멘털 팩터 모델을 사용하던 경제적 팩터 모델을 사용하던 별차이가 없다.  
→ 펀더멘털 팩터 모델이 올바르지 않을 경우 경제적 팩터모델이 더 강력한 이론적 기반이 있기 때문에 경제적 팩터 모델의 사용이 최선  

## 3.4 THE SCREENING AND RANKING OF STOCKS WITH THE Z-SCORE  
많은 포트폴리오 매니저들이 Z-스코어를 사용하여 주식을 선별하고 순위를 매긴다.  
→ 이들은 보다 자주 팩터 모델의 대안으로 사용된다.  

주식 스크리닝 및 순위화에서 Z스코어는 주식의 펀더멘털에 대한 표준화된 노출을 의미한다.  
$z_i = \frac{\beta_i-\mu}{\sigma}$  
ex. 만약 $z_i$가 2라면, 주식 i의 P/B비율은 평균으로부터 2 표준편차 떨어져 있다.  


## 3.5 HYBRIDS OF THE MODELS AND THE INFORMATION CRITERION  
실무자들은 종종 기본적인 QEPM모델에 특별한 모델을 결합하여 추가적인 입력을 시도한다.  
이렇게 만들어진 하이브리드 모델들은 때때로 주식 수익률에 대한 합리적인 설명을 제공하지만, 많은 경우에 원치 않는 영향(정보기준 위반 등)을 끼친다.  

## 3.6 CHOOSING THE RIGHT MODEL  

  

- 경제적 팩터 모델은 펀더멘털 팩터 모델보다 정량적 모델이 탄탄한 경제 이론에 기반해야 한다는 QEPM의 다섯번째 원칙을 더 잘 준수한다.   

ex. 소형주와 가치주가 추가 수익률을 내는 것은 소형주와 가치주이기 때문에 특별히 좋은 것이 아니라, 소형주와 가치주이기 때문에 주식의 위험이 증가하기 때문  
→ 사이즈와 밸류를 보는 것보다 주식의 근본적인 위험을 직접 측정하는 모델을 사용하는 것이 더 나음  


- 두 모델 모두 P/B, P/E, size, momentum 등의 팩터들을 다룰 수 있으나 GDP와 인플레이션 같은 거시경제적 팩터들은 경제적 팩터 모델만이 다룰 수 있다.  
→ 포트폴리오 매니저가 모든 유형의 팩터들을 혼합하여 사용하고 싶다면 경제적 팩터 모델이 필수적  

- 실행의 용이성 측면에서는 경제적 팩터 모델보다 펀더멘털 팩터 모델이 선호됨  
- 펀더멘털 팩터 모델은 많은 양의 역사적 데이터 없이 추정될 수 있음  
- 하지만 횡단면 차원에서 볼 때 경제적 팩터 모델에 대한 데이터 요구사항이 더 낮다.  
→ 경제적 팩터 모델이 소수의 주식에 대해 추정될 때에도 그만큼 추정 오류가 상당할 수 있기에 실제로 경제적 팩터 모델도 포트폴리오 구성 과정에서 유용하게 사용하기 위해서는 많은 수의 주식 데이터가 필요하다.  

## 3.7 CONCLUSION   
- QEPM의 주요 모델로 펀더멘털 팩터 모델과 경제적 팩터 모델 설명  
- 두 모델의 추정치는 서로 다르지만 모델 공식 자체는 동일하다.  
- 펀더멘털 팩터 모델이 올바른 모델이라면 두 모델 간의 차이는 없으나 일반적으로는 경제적 팩터 모델 선호  
- 많은 매니저들은 팩터 모델 대신에 주식 스크리닝 및 순위화 방법 활용
