# Chapter 4. Factors and Factor Choice  
팩터에는 다양한 종류가 존재한다.  
- 펀더멘털  
- 기술적  
- 경제적  
- 대안적  

좋은 팩터들은 주식 수익률과의 관계가 안정적이고 지속적일 뿐만 아니라 경제 이론에 의해 설명될 수 있어야 한다.  

※ FUNDAMENTAL FACTORS  
- 주식 성과에 가장 명백한 영향을 주는 것은 회사의 재무 상태  
- 일반적으로 손익 계산서, 재무상태표, 현금 흐름표 등에서 직접 계산된 비율들  

1. Value: 주가순자산비율(PBR)과 주가수익비율(PER)은 주식이 싼지 비싼지 측정  
2. Size: 시가총액은 회사의 규모를 분류하고 규모가 주식 수익률에 미치는 영향 측정  
3. 운영 효율성, 운영 수익성: 경영진이 회사를 얼마나 잘 운영하고 있는지 알려줌  
4. 지급 능력: 회사가 미래의 단기 채무를 충족할 수 있는 능력을 측정  
5. 재무 위험: 부채 대 자본 비율과 이자 커버리지와 같은 비율로 재무 건전성 측정  
6. 기업 활동: 기업 경영진의 결정과 관련이 있거나 다른 카테고리에 반드시 속하지 않는 팩터들을 측정  


### 4.2.1 Valuation Factors  
  
인기 있는 밸류에이션 팩터들의 값   
- 낮은 PER 주식은 위험 조정 기준으로 높은 PER 주식을 능가하는 경향이 있다.  
- 시장이 일반적으로 나쁜 소식에 과잉반응하여 일부 주식의 가격을 지나치게 저평가한다는 것  
- 매니저들은 저평가된 주식을 매수하고 이후 이 정보가 반영될 때 큰 수익을 얻게된다.  

위 표에서 테슬라의 PER은 1203인 반면 JPMorgan은 15.18이다. 이렇게 산업 간 비율의 격차가 크기 때문에 산업 내에서 비교하는게 적절함.   

앞서 소개한 팩터들은 서로 높은 상관관계를 가지기 때문에 매니저는 자신의 선호에 따라 하나를 선택할 수 있다.  
- 일부 매니저들은 손익이 없는 회사의 PER을 계산할 수 없기 때문, 학술 연구에서의 보편성 때문에 PBR을 선호  
- 일부 매니저들은 P/CF 비율을 대신 사용하는데 그들은 현금흐름 측정이 회사 경영진의 조작에 덜 민감하다고 보기 때문   

높은 배당 수익률과 낮은 P/B, P/S, P/CF, P/EBITDA 비율을 가진 회사들이 그 반대인 주식들을 능가하는 경향이 있다.  

- 오히려 P/B 비율이 너무 낮다면, 경영진이 주주들에게 충분한 수익을 창출하지 못하고 주식 가격이 반응하여 하락한 증거일 수 있음  
- 주가수익성장비율(PEG)은 성장하는 좋은 회사를 식별하기 위해 일반적으로 사용되는 또 다른 가치 측정 방법  
→ P/E 비율만으로는 주식이 고평가되어 보일 수 있지만, 만약 주식이 낮은 PEG 비율을 가지고 있다면, 회사의 예상 이익 성장을 고려할 때 높은 P/E 비율이 합리적일 수 있음  

- P/E 비율은 회사의 예상 성장률과 배당 수익률의 합으로 나누어 한 단계 더 조정할 수 있음  
→ 그 결과 얻어지는 비율인 PEGY는 회사의 예상 성장률과 배당 수익률을 모두 고려할 때 주식의 가치가 얼마나 매력적인지를 나타낸다.  

- 배당 수익률은 또한 회사의 성숙도와 성장 전망을 결정하는 데 귀중한 도움을 줄 수 있음  

### 4.2.2 Size Factors  
  
회사의 시가총액, 사이즈라고 불리는 팩터 또한 인기 있는 펀더멘털 팩터이다.   
- 소형주가 장기적으로 대형주를 능가한다고 알려져 있음  
→ 일부 사람들은 소형 회사가 대형회사보다 더 큰 위험을 안고 있다고 믿으며, 그러므로 추가적인 위험 프리미엄이 정당하다고 생각  
- 다른 사이즈 팩터로는 회사의 보통주 자본과 회사의 총 자산이 있음   


### 4.2.3 Operating Efficiency Factors   
  
운영 효율성 팩터들은 단기 및 장기에 걸쳐 회사가 얼마나 효율적으로 운영되는지를 설명하는 팩터  

- 단기 효율성 분석은 재고 회전율(IT)비율을 포함한다.  
→ 재고 회전율 비율은 매출원가를 평균 재고로 나눈 것으로, 제품이 재고에서 매출로 이동하는 속도를 나타냄. 높은 IT비율이 좋은 신호이다.  

- 회사의 장기 효율성을 테스트하기 위한 총 자산 회전율(TAT) 비율은 순 매출을 평균 총 자산으로 나눔  
→ TAT가 높다는 것은 회사가 자산을 최대한 활용하고 있다는 것을 시사한다.  

자본 회전율(ET)과 고정 자산 회전율(FAT)은 회사가 그 특정 자산 클래스를 얼마나 잘 매출액으로 전환하는지를 보여줌  

- 높은 재고 회전율이 일반적으로 좋긴 하지만, 지나치게 빠른 재고 회전은 재고부족으로 인해 고객의 수요를 충족시키지 못할 위험을 증가시킨다.  
→ 특정 사업 분야의 정상 회전율은 시장 경쟁, 수요 및 업계 관행과 요구에 따라 다르다.  
### 4.2.4 Operating Profitability Factors  
  
- 수익성이 높은 회사들이 소유하기에 가치가 있는 회사들  
- 일반적으로 매출 총 이익률(GPM), 운영 이익률(OPM), 순이익률(NPM)의 연간 성장을 살핀다.  
→ 이 세 가지 비율을 산업 내부, 시장 전체, 그리고 시간에 따라 여러 차원에서 검토하는 것이 항상 중요  
- 다른 수익성 팩터들은 총자본수익률(ROTC)과 자기자본순이익률(ROCE)  


### 4.2.5 Solvency Factors  
  
영업활동으로 인한 현금 흐름 비율(CFOR), 현금 비율(CR), 유동 비율(CUR), 당좌 비율(QR)은 모두 회사가 유동성 있는 현금으로 부채를 충당할 수 있는 능력을 측정하는 지표  
- 지급능력 팩터 값이 높은 회사들은 더 많은 유동성을 가지고 있거나 지급능력이 뛰어나며, 파산의 위험이 적다.  
- 많은 포트폴리오 매니저들은 이러한 변수 중 하나 이상을 재정적으로 건전한 주식을 선별하기 위한 주식으로 사용한다.  
- 높은 CUR은 일반적으로 충분한 단기 유동성을 나타내지만, 현금 또는 기타 매매 가능한 대체재의 비효율적인 사용이 없는지 확인하기 위해 재고 회전율(IT)과 자산 수익률(ROA)과 같은 다른 비율과 함께 검토되어야 한다.  
→ CUR이 높더라도 QR이 비교적 낮으면 대부분의 유동 자산이 재고에 묶여 있다는 뜻  

### 4.4.6 Financial Risk Factors  
  
재무 위험의 징후를 통제함으로써, 포트폴리오 매니저들은 판매가 일시적으로 둔화되거나 경제가 침체되었을 때 폭풍을 견딜 준비가 되어 있지 않은 회사들을 식별할 수 있음  

- 다른 회사들에 비해 높은 부채 대 자본(D/E)비율을 가진 회사는 그 핵심사업이 탄탄할지라도 파산에 취약하다.  
- 높은 부채 대 자본(D/E) 비율만으로는 재정적 위험을 의미하지 않지만 높은 D/E비율과 함께 낮은 이자 보상 비율(ICR)은 더 큰 우려의 원인이 되는데, 이것은 회사의 현재 수익이 많은 이자 지불을 커버하지 못할 것임을 의미한다.  
- 가장 이상적인 상태는 부채가 적고 이에 대한 이자 지급을 커버할 충분한 현금을 손에 쥐고 있는 것  

### 4.2.7 Corporate Activity Factors   
주식 매입, 내부자 매수, 연구 개발(R&D) 지출 결정뿐만 아니라 다른 팩터 범주에 깔끔하게 들어맞지 않는 기타 팩터들과 같은 기업 활동에 대한 팩터  
- 대표적으로 R&D 지출은 경영진이 미래에 적극적으로 베팅하고 있다는 것을 드러내며, 이는 유익한 기업 활동의 징후일 수 있음  

## 4.3 TECHNICAL FACTORS  
기술적 팩터들은 대부분 과거의 가격과 거래량 데이터(보통 시가, 고가, 저가, 종가, 거래량, 미결제약정, 매수 및 매도 가격)와 기타 쉽게 구할 수 있는 재무 정보로부터 구성된다.  
- 기술적 팩터의 가장 큰 장점 중 하나는 지속적으로 데이터가 업데이트 된다는 것  
→ 펀더멘털 데이터의 경우 기업이 분기별로 제출하기 때문에 분기별로 업데이트할 수 있지만 기술적 지표는 몇초마다 업데이트가 이루어짐   

기술적 팩터들은 유동성 위험 팩터, 가격 기반 팩터, 거래량 기반 팩터, 그리고 전체 시장 움직임 팩터의 네 가지 하위 범주로 분류  
1. 유동성 위험 팩터: 주식 거래의 결과와 영향을 이해하기 위해 사용  
2. 가격 기반 팩터: 주로 주식 가격이나 수익률에서 생성되며 주식의 잠재적 미래 가격 움직임을 나타내는 지표로 사용  
3. 거래량 기반 팩터: 과거 거래나 거래량의 정보에서 주로 생성되며 시장 참여자들의 의도와 변화하는 행동을 신호할 수 잇음  
4. 전체 시장 움직임 팩터: 전체 주식시장 움직임을 해독하고 가까운 미래에 대한 그들의 함의를 이해하는 데 도움이 될 수 있는 종합 기술적 지표    

### 4.3.1 Liquidity Risk Factors  
  
위 표는 거래 회전율(TT), 애미허드 비유동성, 불변성 비유동성과 같은 유동성 팩터를 나타낸다.  
- 유동성 팩터들은 주식의 유동성이 거래 수요와 일치하는지를 측정한다.  
→ 가장 쉽고 흔한 유동성 요인 중 하나는 거래회전율(TT). 평균 일일 달러 거래량(ADDTC)을 시가총액으로 나눈 값으로, 평균적으로 하루에 거래되는 상장 주식의 백분율 측정  

- 애미허드 비유동성(AILIQ)은 주식의 일일 수익률을 일일 거래량으로 나눈 비율에 기반한 비유동성 척도  
→ 주어진 거래량에 비해 일일수익률이 높을수록 주식의 유동성이 가격 영향으로 인해 낮다는 개념을 바탕으로 한다.  

- 비유동성에 대한 더 최근 척도는 불변성 비유동성(INVIL)으로 이는 이론적 모델에 기반한 비유동성을 측정  
→ 주식의 평균 달러 거래량과 그 변동성과 관련이 있음  

이러한 모든 유동성 팩터들은 매니저의 포트폴리오 크기와 체결하고자 하는 거래의 가치와 관련해서만 중요하다.  
→ 거래 유동성은 매니저가 시장에서 가능한 주식 수 이상을 매수 or 매도하고자 할 때 고려된다.  
일부 포트폴리오 매니저들은 포트폴리오 모델에 유동성 팩터를 거래비용으로 포함시키는 반면, 유동성 제약을 고려하여 트레이딩 데스크에 맡겨 거래비용을 정리하는 경우도 존재  

### 4.3.2 Price-Based Factors   
  
이러한 팩터들은 주식 가격의 단기 움직임이나 미래의 주식 수익률에 대해 무언가를 나타낼 수 있는 주식 가격과 관련된 측정치를 포착하려고 시도   

- 여러 학술 연구에서 주식 수익률의 양의 자기상관성을 발견했다. → 모멘텀  
- 개별 주식 수익률에 대해 일주일과 한 달 기간에 걸쳐 음의 자기상관성의 증거도 있음  
→ 일부 포트폴리오 매니저들은 지난 해의 주식 수익률로 모멘텀을 측정하고 일부는 지난 달의 주식 수익률로 측정하며 더 복잡한 변형도 이루어진다.  

- 상대 강도 지수(Relative Strength Index, RSI)는 주식을 매수할 때인지 매도할 때인지를 정확히 파악하려는 인기 있는 지표  
- 볼린저 밴드는 매수와 매도를 위한 평균 가격 주변의 한계를 나타내며, 단순 이동 평균 모델을 개선  
→ 주식이 평균 위의 상단 밴드에 도달하면 매도, 평균 아래의 하단 밴드에 도달하면 매수 신호이다.  

- 베타는 또한 각 주식의 베타 노출이 회귀분석을 통해 추정되기 때문에 다른 팩터들과 다름  
→ 1970년대 초반부터의 연구는 베타가 미래주식 수익률을 예측하는 데 유용할 수 있음을 보여주었다. 하지만 이론과는 달리, 사람들은 저베타 주식이 고베타 주식보다 더 높은 위험 조정 수익률을 가지고 있다는 것을 발견했다.  

### 4.3.3 Volume-Based Factors  
  
위 표는 공매도 비율을 나타낸다.  
- 어떤 사람들은 높은 공매도 비율을 반대 신호로 사용  
→ 많은 투자자들이 주식을 공매도하게 되면 주식의 가격은 점점 오르고 그 주식을 공매도한 투자자들이 주식을 다시 매수해야만 하는 상황에 처함. 이는 주식을 더욱 상승시키는 결과로 이어진다.  

ex. 2015년 12월에 테슬라의 공매도 비율은 64였고 많은 사람들은 테슬라의 주가가 과대평가되었다고 생각했지만 실제로 그 이후 테슬라 주식은 18배 상승하였다.  


## 4.4 ECONOMIC FACTORS

- 이러한 변수들은 각각 다른 방식으로 정의될 수 있지만, 일부 표준 정의들이 존재  
→ 퀀트 매니저는 거시경제 수치를 계산할 때 올바른 데이터를 사용해야 한다.  

- 거시경제 데이터의 측정과 발표 사이에는 상당한 지연이 존재하기 때문에 정보의 지연을 고려하여 모델을 조정해야 한다.

- 애널리스트 권고는 1에서 5까지의 숫자로 제공되며, 1은"강력 매수", 2는 "매수", 3은 "보유" 또는 "중립", 4는 "매도", 5는 "강력 매도"를 의미   
→ ex. 애널리스트 권고 중앙값(ARMR)에서 테슬라는 3으로 '보유'를 의미  

만약 애널리스트에 의해 커버되지 않는 주식은?  
→ 오히려 이점이 될 수 있다. 소외된 기업 효과에 따르면 소수의 애널리스트만이 추적하는 주식은 대량으로 커버된 주식보다 장기간 위험조정 수익률이 상대적으로 높은 경향이 존재  
- 소외된 기업 효과는 좋은 뉴스가 있을 때 소외된 기업을 매수하고 나쁜 뉴스가 있을 때 매도하는 전략  
→ 동일한 기본 조건을 가진 두 주식이 있다면, 모델은 잘 알려진 주식보다 소외된 주식을 더 높게 평가할 것   


### 4.5.2 Captivus Factors

→ 이 정보를 캡쳐하는 것이 매우 가치있다.  

- 이러한 유형의 데이터는 종종 대체데이터 또는 빅데이터라고도 불림  
- Captivus 데이터는 일반적인 기본 데이터보다 더 일찍 얻어짐.
- 또다른 이점은 Captivus 데이터가 일반적으로 보다 전통적인 데이터 소스를 보완하거나 상호 보완적이다.  
- Captivus 데이터를 얻는 가장 인기있는 방법은 웹 스크래핑   
- Captivus 데이터는 전통적인 데이터보다 다양하고 종종 비싸고 얻기 어려움  
→ 하지만 정보의 더 높은 빈도, 정확성 등에 있어 퀀트 투자자들에게 더 많은 이점을 제공할 수 있기에 미래에 QEPM 데이터 도구 키트의 중요한 부분이 될 것

## 4.6 FACTOR CHOICE  
앞에서 정리한 팩터들을 조화롭게 결합하여 투자모델을 구축해야 한다.  
- 퀀트 패니저는 팩터를 적절히 선택하고 결합하기 위해 재무 및 경제 이론을 이해해야하며 QEPM의 일곱가지 원칙을 염두해두고 있어야 한다.  

1. 펀더멘털 팩터 모델을 위한 팩터를 선택하기 위해 단변량 및 다중회귀 기법을 사용할 것을 제안   
2. 경제적 팩터 모델을 위한 팩터를 선택하기 위해 단일 차원 및 다차원 제로투자 포트폴리오 기법 사용 제안  

- 이외에도 단순 상관관계 또는 켄달의 순위 상관관계를 사용하여 팩터 선택 간의 상관관계를 결정할 수 있음  

### 4.6.1 Univariate Regression Tests   
- 이론적으로 주식의 수익률을 설명할 수 있다고 정당화할 수 있는 팩터들의 그룹을 식별  
- 이후 유니버스 내의 주식 수익률에 대해 각 팩터에 대한 패널 회귀 분석 시행  
$r_{i,t} = \alpha + f\beta_{i,t} + \varepsilon_{i,t}$  
단순선형회귀의 단점은 포트폴리오 매니저가 주식 수익률을 설명하는 데 중요한 많은 변수들을 찾을 수는 있지만 그 변수들이 서로 대체재가 될 수 있다는 것  
→ ex. P/E, P/B 처럼 상관관계가 높은 팩터 중복 가능  
그럼에도 불구하고 단순회귀 사용의 장점?  
→ 수행하기 매우 간단하며 가능성있는 팩터들이 너무 많을 경우 초기 선별에 활용 용이  


### 4.6.2 Multiple Regression Tests  
다중 회귀 분석은 기대 수익률에 미치는 영향을 추정하면서 팩터들 간의 관계를 고려하기 때문에 어떤 팩터들을 포함할지 결정하기에 더 정교한 방법  
$r_{i,t} = \alpha + f_1\beta_{i,1,t} + \ldots + f_K\beta_{i,K,t} + \varepsilon_{i,t}$  
→ 데이터 마이닝 과정에서의 문제를 고려해주어야한다. 

- 팩터를 선택하기 위한 최선의 조언은 특정 팩터들을 가진 모델이 작동할 이유에 대한 강력한 이론적 설명으로 시작하는 것  
- 모델의 견고성을 위해 다양한 기간에 걸쳐 모델을 테스트해야 함   


### 4.6.3 Unidimensional Zero-Investment Portfolio  
포트폴리오 매니저들은 특정 팩터를 기반으로 제로 투자 포트폴리오를 구성하고 그 특성을 연구할 수 있다.  

- 최상위 그룹의 수익률을 마지막 그룹의 수익률에서 뺀 것이 제로 투자 포트폴리오의 수익률  
- 제로 투자 포트폴리오 수익률이 계산된 후에는 평균 포트폴리오 수익률이 0과 유의미하게 다른지에 대한 통계적 검정  

### 4.6.4 Multidimensional Zero-Investment Portfolio  
여러 팩터를 동시에 고려하여 제로 투자 포트폴리오를 생성할 수도 있음  

- 포트폴리오 매니저는 관심 있는 각 팩터별로 모든 주식의 순위를 매긴다.  
- 만약 작은 사이즈와 낮은 P/B 비율이 함께 주식 수익률에 영향을 미치는지에 관심이 있다면, 작고–낮은 포트폴리오에 롱 포지션을 취하고, 크고–높은 포트폴리오에 숏 포지션을 취함으로써 제로 투자 포트폴리오를 만들 수 있음  
- 이후 t-통계값을 계산하여 팩터들의 공동 효과가 유의미한지를 결정   

또는 다중회귀분석을 적용할 수도 있다.  
- 고려하는 각 팩터에 대해 주식이 최상위 그룹에 속하면 1, 최하위 그룹에 속하면 -1, 그 외에는 0의 값을 가지는 지표 변수 정의  
- 주식 수익률에 대한 지표 변수의 회귀를 실행하면 추정계수는 곧 제로 투자 포트폴리오의 수익률로 해석될 수 있다.  

### 4.6.5 Techniques to Reduce the Number of Factors   
이번 절에는 간단한 상관관계에 기반한 통계와 순위 상관관계에 기반한 통계 두 가지 유형에 대해 논의  

- 첫번째로는 '내부 상관계수(within-correlation coefficient)  (Q.피어슨 상관계수?)  
$\hat{\rho}(X_{i,t}, Y_{i,t}) = \frac{\hat{C}(X_{i,t} - \bar{X}_t, Y_{i,t} - \bar{Y}_t)}{\hat{S}(X_{i,t} - \bar{X}_t) \hat{S}(Y_{i,t} - \bar{Y}_t)}$  
$\hat{C}, \hat{S}$는 각각 샘플 공분산과 표준편차이며, $\bar{X}_{t}, \bar{Y}_{t}$는 $t $ 시점의 $X$와 $Y$의 샘플 평균  

- 두번째로는 켄달의τ  
켄달의 $τ$는 순위 상관관계(한 변수의 순위화가 다른 변수의 순위화와 관련이 있는지)를 보여줌  

- 두 관측값을 선택한다고 가정  
- 두 관측값을 한 변수의 값에 따라 순위를 매김  
- 그렇다면 한 변수에 의한 순위가 다른 변수에 의한 순위와 동일할 확률은?  
→ 1이면 모든 관측값 쌍에 대해 한 변수에 의한 순위가 다른 변수에 의한 순위와 동일, -1이면 순위가 완전히 반대, 0이면 절반은 순위가 같고 절반은 반대  

★ 켄달의 순위 상관관계는 값의 크기차이를 반영하지 않고 오로지 순위로만 판단. 팩터값의 크기가 얼마나 더 크냐가 그렇게 중요한 것이 아니기 때문에 상대적인 순위로 판단한다.   → 크기가 클수록 좋은 팩터: 모멘텀 팩터. 수익률은 무조건 클수록 좋기 때문에 이 경우에는 크기 비교가 이루어져야 한다.

## 5.1 INTRODUCTION   

주식 스크리닝은 투자 유니버스 내의 주식들을 순위를 매겨 매니저의 관점에 따라 더 유리하거나 덜 유리한 주식으로 쉽게 구분할 수 있게 하는 것  
또한 투자 유니버스를 좁히는 데에도 사용가능  

- 순차적 스크리닝과 동시적 스크리닝이라는 두가지 종류가 존재  

순차적 스크리닝: 포트폴리오 매니저가 주식 선정 기준을 우선순위화  
- 가장 중요한 기준으로 먼저 유니버스 내의 주식들을 제거, 이후 두번째로 중요한 기준에 따라 반복    

주식 스크리닝의 더 발전된 버전은 멀티팩터를 활용한 동시적 스크리닝  
→ 모든 기준을 한꺼번에 적용하여 종합 점수에 따라 순위가 매겨진다.  

- 가장 인기 있는 동시적 스크리닝 방법 중 하나는 종합 Z-score접근법  
→ 만약 매니저가 이미 펀더멘털 혹은 경제적 팩터 모델을 사용하고 있다면, 특정 조건이 충족될 때 Z-스코어를 모델의 상수항 α로 추가할 수 있음

피터 린치, 워렌 버핏, 필립 피셔 등 그들만의 다양한 스크리닝 전략으로 많은 수익률을 얻을 수 있었음.

## 5.4 SIMULTANEOUS SCREENING AND THE AGGREGATE Z-SCORE  
- 투자 대상을 더 정교하게 선정하는 방법은 멀티팩터를 이용한 동시 선별을 사용하는 것  
- 동시 선별에서는 포트폴리오 매니저가 여러 팩터에 따라 주식을 선택하지만, 한 번에 한 팩터씩이 아니라 모든 팩터를 동시에 선별  
- 동시적 선별의 장점은 순차적 선별 과정에서 생기는 부족한 주식 양의 문제가 일어나지 않는다는 것  

어떻게 한번에 모든 팩터를 선별하는가?  
→ 다른 유형의 팩터들이 다른 단위로 표현되기 때문에, 모든 팩터 값을 더하는 것은 X  
따라서 표준화 or 정규화 작업을 통해 변수들의 단위를 같게 만들어준다.  


### 5.4.1 The Z-Score  
위의 표준화 과정에서 가장 많이 사용되는 방법 중 하나가 Z-score 방법.  
→ 팩터 값들을 표준화해줌으로써 특정 관측값이 평균에서 얼마나 떨어져있는지 알 수 있음  
Z-score 별로 정렬할 경우 PBR과 같이 값이 낮을수록 좋은 팩터들의 경우 B/P 비율을 사용하는 것이 바람직  
$Z_{i,k} = \frac{\beta_{i,k} - \overline{\beta}_k}{S(\beta_k)}$  

### 5.4.2 The Aggregate Z-Score  
위 과정을 거치면 이젠 단순히 모든 값들을 더해주면 된다.  
→ 각 주식별로 모든 팩터의 z-score를 더한 값이 곧 그 주식의 종합 z-score  
- 일부 개별 팩터 또는 개별 주식의 종합 z-score는 극단적인 값을 가질 수 있다.  
→ z-score 값이 이상치로 판단된다면 3이상일 경우 3으로, -3 이하일 경우 -3으로 조정  
- 극단적인 z-score값을 가진 주식을 배제하는 방법도 존재하지만 이는 잠재적으로 매우 높은 수익률 or 매우 낮은 수익률을 가진 주식에 대한 좋은 정보를 잃을 수도 있음  

- 주식의 종합 z-score 내에서 각 팩터를 어떻게 가중할지 선택할 수 있다.  
- 일부 매니저들은 동일 가중치 방식을 활용하는데 이는 비교적 더 안정적인 결과를 생산하기 때문  


### 5.4.3 Ad Hoc Aggregate Z-Score  
- 이외에도 포트폴리오 매니저의 투자 스타일에 따라 팩터들을 가중할 수 있음  
- 일부 매니저들은 각 팩터에 대해 십분위or오분위 포트폴리오를 만들고 팩터의 역사적 정보비율을 계산하여 얻은 팩터의 정보비율에 따라 가중  
→ 팩터 간 상관관계의 영향을 간과할 수 있음  


### 5.4.4 Optimal Aggregate Z-Score  
- Ad Hoc 방법의 가중은 팩터의 인지된 중요성에 따라 가중치를 달리 할당하려고 시도하지만, 이러한 방법들은 매우 주관적  
- 동일 가중과 Ad Hoc 방법 모두 팩터 간의 상관관계를 고려하지 않는다.  

- 최적 가중치의 매우 일반적인 접근법 중 하나는 역사적 샘플 데이터 세트를 사용하여 통계학적 방법으로 최적 노출을 추정하는 것  
→ 유니버스에 있는 모든 주식의 월간 수익률 시리즈와 각 주식에 대한 월 초의 팩터 Z-sore 값을 취합하여 샘플 기간 동안 횡단면 회귀 분석을 실행하여 최적 Z-스코어 가중치를 찾는다.    

$r_{i,t} = \gamma_i + \delta_1 Z_{i,1,t-1} + \delta_2 Z_{i,2,t-1} + \ldots + \delta_K Z_{i,K,t-1} + \epsilon_{i,t}$  
$\gamma_i$는 상수항, $\delta_k$는 주식 수익률에 대한 팩터 Z-스코어 $k$의 기여도에 대한 계수 추정치, $\epsilon_i$는 오차항  
- 포트폴리오 매니저는 세 가지 이상의 가상 경제 환경을 구성하고 각각에 대한 최적의 Z-스코어 가중치를 계산  


### 5.4.5 Factor Groups and the Aggregate Z-Score  
이러한 정량적 포트폴리오 매니저들이 사용하는 K개 팩터들을 M개의 팩터 그룹으로 나눌 수 있음  ex. P/B + P/E + S/P : 가치 평가 그룹  

이러한 과정이 가지는 이점  
- 선별 과정을 조직화할 수 있음  
→ 모델의 수정과 투자 위원회에 성과 결과를 발표하는 것을 단순화함  
- Value를 기준으로 할 경우 P/B만 사용하는 것보다 가치평가그룹 팩터를 사용하는 것이 효율적  
- 경제 상황이 변화함에 따라 팩터 그룹에 중요도를 쉽게 변경할 수 있음  

## 5.5 THE AGGREGATE Z-SCORE AND EXPECTED RETURN   

### 5.5.1 Expected Return Implied by the Z-Score  
종합 Z-스코어를 사용하여 실제 주식 수익률에 대한 회귀분석을 수행함으로써 예상 주식 수익률을 구할 수 있음   
$r_{i,t} = \gamma_i + \delta Z_{i,t-1} + \varepsilon_{i,t}$  
$E(r_{i,T+1}) = \gamma_i + \delta Z_{i,T}$  

  

하지만 이 방법론에는 몇가지 문제점이 있다.  
1. Z-score는 기간 간에 변화가 크지 않을 수 있지만 팩터 프리미엄은 크게 변할 수 있다.  
→ 추정계수가 안정적이거나 유의하지 않을 확률이 높음. 이러한 문제를 줄이는 방법으로는 더 긴 기간의 역사적 데이터에 대한 패널 회귀를 실행하는 것.  
2. Z-score가 향후 수익률 간의 상관관계가 낮을 수 있다.  
→ t시점의 Z-score 값과 t+1시점의 주식수익률 간의 회귀분석은 엄격한 이론에 기반하지 않기 때문에 연결이 약할 수 있다.  
3. 기존의 종합 Z-score 방법보다 비교적 복잡하다.  



### 5.5.2 Forecasting Rule of Thumb  

회귀분석 식을 변형한 식  
- 회귀를 실행하는 것보다 IC와 변동성을 계산하는 것이 더 쉬운 경우에 유용  

### 5.5.3 The Equivalence between the Z-Score Model and the Fundamental Factor Model  
특정 조건 하에서 Z-score 모델이 펀더멘털 모델과 동일한 기대 수익률을 생성한다는 것을 보여줄 수 있다.  
- 특히 팩터 프리미엄이 팩터 익스포져의 표준편차에 반비례한다면, 종합 Z-score 모델은 펀더멘털 팩터 모델과 동일한 결과를 도출  

팩터 프리미엄이 팩터 익스포져의 표준편차 값에 반비례한다고 가정  
$f_k = \frac{c}{\sqrt{V(\beta_k)}}$  
따라서 펀더멘털 팩터 모델은 다음과 같이 쓸 수 있다.  
$r_i = \alpha_i + \beta_{i,1}f_1 + \cdots + \beta_{i,K}f_K + \epsilon_i \\= \alpha_i + c\frac{\beta_{i,1}}{\sqrt{V(\beta_1)}} + \cdots + c\frac{\beta_{i,K}}{\sqrt{V(\beta_K)}} + \epsilon_i \\= \tilde{\alpha}_i + cZ_i + \epsilon_i$  

## 5.6 THE AGGREGATE Z-SCORE AND THE MULTIFACTOR α  
일반적으로 많은 포트폴리오 매니저들은 벤치마크 대비 최적화를 수행하기 위해 상용 소프트웨어를 사용  
→ 이를 통해 사용자가 멀티팩터 알파를 제공할 수 있게 한다.  

많은 포트폴리오 매니저들은 종합 Z-score를 멀티팩터 알파로 변환  
→ 가장 간단한 방법은 실제 Z-score를 알파로 사용하는 일대일 변환  
- 문제는 Z-score가 실제 수익률을 대표하지 않는다면 실제 초과 수익률 잠재력에 의해 정당화되는 것보다 훨씬 더 높은 예측 Z-score를 가진 주식을 선호함으로써 주식 선택 과정이 왜곡될 수 있음  

두번째 방법은 $E(r_{i,T+1}) = \gamma_i + \delta Z_{i,T}$ 의 기대 초과 수익률을 알파로 사용하는 것  

세번째 방법은 모든 주식의 팩터 익스포져와 팩터 프리미엄을 구하여 역사적 알파를 계산한 후 알파와 종합 Z-score 사이의 관계를 추정  
$r_{i,t} - \hat{\beta}_i'f_t = \gamma_i + \delta z_{i,t-1} + \epsilon_{i,t}$   
→ 매니저는 이 회귀에서 얻은 추정치를 사용하여 다음 시점의 모든 주식의 알파를 예측할 수 있음  

하지만 이 세가지 방법 모두 단점 존재  
- Z-score를 계산하는데 사용한 데이터와 팩터 모델을 계산하는데 사용한 데이터가 충분히 다르지 않으면 심각한 왜곡이 발생할 수 있음  
- 두번째와 세번째 방법은 계산과정이 더 복잡  
- 세번째는 더 많은 추정 기간과 시간이 지남에 따라 추정치의 안정성이 높아져야 유의미함
