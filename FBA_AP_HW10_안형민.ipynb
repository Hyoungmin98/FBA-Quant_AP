```python
# 4. 12
# (a): Compute the mean, upper, and lower bands of a
# Bollinger band with two standard deviations (i.e., l =2) and a 10-day rolling window (i.e., k = 10) for the
# dates July 26, 2020July 29, 2020.

import pandas as pd

# AALC stock data from the image
data = {
    "Date": [
        "7/13/2020", "7/14/2020", "7/15/2020", "7/16/2020", 
        "7/19/2020", "7/20/2020", "7/21/2020", "7/22/2020", 
        "7/23/2020", "7/26/2020", "7/27/2020", "7/28/2020", "7/29/2020"
    ],
    "Price": [
        11.31, 11.41, 11.28, 11.57, 11.43, 11.41, 10.92, 
        10.51, 10.35, 9.35, 9.65, 10.01, 9.67
    ]
}

# Creating a DataFrame
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Calculating Bollinger Bands with a 10-day rolling window and 2 standard deviations
window = 10
df['Rolling Mean'] = df['Price'].rolling(window=window).mean()
df['Rolling Std'] = df['Price'].rolling(window=window).std()
df['Upper Band'] = df['Rolling Mean'] + (2 * df['Rolling Std'])
df['Lower Band'] = df['Rolling Mean'] - (2 * df['Rolling Std'])

df
```

```python
# (c), (d)

# Calculate daily price changes
df['Change'] = df['Price'].diff()

# Separate gains and losses
df['Gain'] = df['Change'].apply(lambda x: x if x > 0 else 0)
df['Loss'] = df['Change'].apply(lambda x: -x if x < 0 else 0)

# Define Wilder period
period = 9

# Calculate initial average gain and loss
df['Avg Gain'] = df['Gain'].rolling(window=period, min_periods=period).mean()
df['Avg Loss'] = df['Loss'].rolling(window=period, min_periods=period).mean()

# Apply Wilder's smoothing for average gain and average loss
for i in range(period, len(df)):
    if i > period:
        df.loc[df.index[i], 'Avg Gain'] = (df.loc[df.index[i-1], 'Avg Gain'] * (period - 1) + df.loc[df.index[i], 'Gain']) / period
        df.loc[df.index[i], 'Avg Loss'] = (df.loc[df.index[i-1], 'Avg Loss'] * (period - 1) + df.loc[df.index[i], 'Loss']) / period

# Calculate RS and RSI
df['RS'] = df['Avg Gain'] / df['Avg Loss']
df['RSI'] = 100 - (100 / (1 + df['RS']))

df
```

Q. A portfolio manager has the data available in the following table. She believes in using a factor group approach to creating aggregate Z-scores. She has decided to construct four groups (i.e., M = 4). Group 1 is titled the valuation group (V), group 2 is titled the technical group (T), group 3 is titled the profitability group (P), and group 4 is the alternative group (A). She also has seven factors that she has identified in predicting stock returns. They are the book-to-price (B/P) ratio, the earnings-to-price (E/P) ratio, the cash-flow-to-price (CF/P) ratio, the 12-month past stock return (M12M), standardized unanticipated earnings (SUE), current accruals (CACC), and net profit margin (NPM).    


(a):  In which group would each of these factors most likely
 be placed?  
 
A.  
(B/P), (E/P), (CF/P): valuation group (V)   
12  (M12M): technical group(T)  
(NPM), current accruals(CACC): profitability group (P)    
   (SUE): alternative group (A)  










```python
# (b): Compute the Z-score of each factor and each stock  

import numpy as np
import pandas as pd

# Data from the image
data = {
    "Ticker": ["CL", "KMB", "PG", "EL", "CLX", "CHD"],
    "B/P": [0.009, 0.007, 0.140, 0.074, 0.044, 0.148],
    "E/P": [0.037, 0.051, 0.040, 0.010, 0.045, 0.036],
    "CF/P": [0.051, 0.082, 0.052, 0.047, 0.065, 0.048],
    "M12M": [27.159, 0.999, 14.150, 29.777, 34.485, 25.475],
    "SUE": [0.500, 3.667, 3.750, 8.000, 2.133, 1.000],
    "CACC": [-0.065, -0.089, -0.039, -0.043, -0.078, -0.038],
    "NPM": [0.167, 0.125, 0.189, 0.044, 0.161, 0.164]
}

# Mean and standard deviation values for each factor from "Universe"
mean_values = {
    "B/P": 0.275, "E/P": 0.041, "CF/P": 0.080, "M12M": 9.095, 
    "SUE": 2.067, "CACC": -0.045, "NPM": 0.108
}
std_values = {
    "B/P": 0.339, "E/P": 0.026, "CF/P": 0.026, "M12M": 17.009, 
    "SUE": 3.017, "CACC": 0.026, "NPM": 0.082
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate Z-scores for each factor
for factor in mean_values.keys():
    df[f"{factor}_Z"] = (df[factor] - mean_values[factor]) / std_values[factor]

# Display the resulting Z-scores
z_scores = df[['Ticker'] + [f"{factor}_Z" for factor in mean_values.keys()]]
z_scores
```

```python
# (c) Suppose that factors within each group are equally weighted. Compute the aggregate Z-score for every factor group. 

# Compute equal-weighted aggregate Z-scores for each group
df['Z_V'] = df[['B/P_Z', 'E/P_Z', 'CF/P_Z']].mean(axis=1)
df['Z_T'] = df[['M12M_Z']].mean(axis=1)
df['Z_P'] = df[['NPM_Z', 'CACC_Z']].mean(axis=1)
df['Z_A'] = df[['SUE_Z']].mean(axis=1)

# Display the results
z_scores = df[['Ticker', 'Z_V', 'Z_T', 'Z_P', 'Z_A']]
z_scores
```

```python
# (d) After careful analysis, the portfolio manager has found that the optimal weights of the factor groups are 20%, 30%, 20%, and 30%, respectively, for the four groups. Compute the aggregate Z-score for each stock.

df['Aggregate_Z'] = (df['Z_V']*0.2 + df['Z_T']*0.3 + df['Z_P']*0.2 + df['Z_A']+0.3)

aggregate_Z_scores = df[['Ticker', 'Aggregate_Z']]
aggregate_Z_scores
```
