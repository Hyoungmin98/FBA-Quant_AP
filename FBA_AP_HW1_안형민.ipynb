## 5.4 The Normal Distribution  
1927-2021년 동안의 Broad Market Index의 월별 수익률 빈도 분포가 정규 분포 형태를 띄고 있다는 걸 알 수 있음.  
  

수익률이 정규 분포 형태로 근사할 수 있을 때의 이점  
1. 대칭성: 정규 분포는 대칭적이기 때문에 평균을 기준으로 양의 편차와 음의 편차의 확률이 동일. 대칭성이 없으면 표준 편차는 위험을 불완전하게 측정하게 된다.  

2. 안정성: 정규 분포는 "안정적"인 분포로, 정규 분포를 가진 자산들이 포트폴리오를 구성할 때 포트폴리오의 수익률도 정규 분포를 따른다.  

3. 두 개의 매개변수: 정규 분포는 평균과 표준 편차라는 두 개의 매개변수만으로 완전히 특징지어지므로, 가능한 시나리오에 대한 확률 할당이 단순해진다.  

4. 상관 관계: 증권의 수익률 간 통계적 의존성을 고려해야 하는데, 정규 분포에서는 상관 계수 하나로 그 관계를 요약할 수 있다.  



## 5.5 Deviations from Normality and Tail Risk  
$\text{Skew} = \text{Average} \left[ \frac{(R - \overline{R})^3}{\hat{\sigma}^3} \right]$  
음의 왜도 값을 가지는 분포를 '왼쪽으로 치우친 분포', '왼쪽 꼬리가 두꺼운 분포'라고 부르며 이는 왼쪽에 극단적인 데이터가 많이 존재한다는 것을 의미  
→ 수익률 분포가 음의 왜도 값을 가지면 극단적인 음값 데이터가 많아지면서 표준편차가 리스크를 과소평가하게됨.  
$\text{Kurtosis} = \text{Average} \left[ \frac{(R - \overline{R})^4}{\hat{\sigma}^4} \right] - 3$  
마찬가지로 첨도가 크면 극단적인 데이터를 과소평가하게됨    

※ Value at Risk  
- 특정 기간 동안 특정 신뢰 수준에서 자산 또는 포트폴리오의 최대 잠재 손실을 나타낸다. 일반적으로 99% 신뢰수준 활용  
$ VaR(1\%, normal) = Mean − 2.33SD$  


※ Lower Partial Standard Deviation and the Sortino Ratio   
수익률 분포가 비정규 분포일 때 표준편차를 위험 척도로 사용할 경우 발생하는 문제들을 방지하기 위해 표준편차를 LPSD로 대체.   
$LPSD = \sqrt{\frac{1}{n}\sum^n_{i=1}min(0,R_i - R_f)^2}$

→ 음값의 수익률 데이터만을 활용하여 '왼쪽 꼬리 표준편차'를 사용, 표본 평균이 아닌 무위험 수익률으로 초과수익률을 계산하여 제곱한 값을 평균. 이 지표를 sharpe ratio에 대입한 것을 Sortino ratio라고 한다.  


※ Arithmetic Average & Geometric Average Return  
1. Arithmetic Average  
  

2. Geometric Average  
$Terminal \, value = (1+r_1)(1+r_2)...(1+r_n)= (1+g)^n  $  
$g = Terminal\, value^{\frac{1}{n}} - 1$    

$ E[Geometric \, average] = E[Arithmetic \, average] - \frac{1}{2}\sigma^2$  
- 수익률의 변동성이 클수록 산술평균과 기하평균의 차이가 커진다.

## Lesson5. Market Mechanics  

매수 호가 (Bid Price): 시장에서 구매자가 특정 자산을 사기 위해 제시하는 최고 가격  
매도 호가 (Ask Price): 시장에서 판매자가 특정 자산을 팔기 위해 제시하는 최저 가격  
매수매도 스프레드 (Bid-Ask Spread): 매도 호가와 매수 호가의 차이  
매수매도 스프레드 = 매도 호가 − 매수 호가  

유동성(Liquidity): 자산을 시장 가격에 큰 변동 없이 신속하게 현금으로 전환할 수 있는 능력을 의미. 유동성이 높은 시장에서는 자산을 쉽게 사고 팔 수 있으며, 유동성이 낮은 시장에서는 자산을 거래하는 데 시간이 걸리거나 시장 가격에 큰 변동이 생길 수 있다.  

틱 데이터(Tick Data): 금융 시장에서 발생하는 각 개별 거래나 가격 변동을 실시간으로 기록한 데이터.  
구성요소: 시간(timestamp): 거래가 발생한 정확한 시간, 거래 가격(trade price), 거래량(volume), 매수호가(bid price), 매도호가(ask price), 매수호가 수량(bid size), 매도호가 수량(ask size)

```python
# Quiz 2
import pandas as pd

def days_to_weeks(open_prices, high_prices, low_prices, close_prices):
    """Converts daily OHLC prices to weekly OHLC prices.
    
    Parameters
    ----------
    open_prices : DataFrame
        Daily open prices for each ticker and date
    high_prices : DataFrame
        Daily high prices for each ticker and date
    low_prices : DataFrame
        Daily low prices for each ticker and date
    close_prices : DataFrame
        Daily close prices for each ticker and date

    Returns
    -------
    open_prices_weekly : DataFrame
        Weekly open prices for each ticker and date
    high_prices_weekly : DataFrame
        Weekly high prices for each ticker and date
    low_prices_weekly : DataFrame
        Weekly low prices for each ticker and date
    close_prices_weekly : DataFrame
        Weekly close prices for each ticker and date
    """
    
    open_prices_weekly = open_prices.resample('W').first()
    
    high_prices_weekly = high_prices.resample('W').max()
    
    low_prices_weekly = low_prices.resample('W').min()
    
    close_prices_weekly = close_prices.resample('W').last()
    
    return open_prices_weekly, high_prices_weekly, low_prices_weekly, close_prices_weekly

import quiz_tests
quiz_tests.test_days_to_weeks(days_to_weeks)
```
