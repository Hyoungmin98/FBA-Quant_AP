### 8.1 A Single-Factor Security Market  

$r_i = E(r_i) + unanticipated  \, surprise$  
→ 공통적으로 적용되는 시장 불확실 변수m과 기업별 불확실 변수 e_i 로 분류  
m과 e_i는 상관관계가 없는 것으로 가정한다. e_i는 기업에 따라 다르기 때문에 전체 경제에 영향을 미치는 공통 요인에 대한 충격과는 무관  

$r_i = E(r_i) + \beta_im + e_i$  
$\sigma^2_i = \beta^2_i\sigma^2_m + \sigma^2(e_i)$    
$Cov(r_i,r_j) = \beta_i\beta_j\sigma^2_m$  
$Corr(r_i,r_j) = Corr(r_i,r_M) * Corr(r_j,r_M)$   
(m: market factor, $e_i$: firm-specific surprise, $\beta_i$:sensitivity coefficient for firm i, )  


### 8.2 The Single-Index Model  
$R_i(t) = \alpha_i + \beta_iR_M(t) + e_I(t)$  
($\alpha_i$: 시장 초과수익이 0일 때 주식의 예상 초과수익)  
- 공격적인 주식은 평균보다 높은 민감도를 가지고 있기 때문에 베타값이 1보다 큰 반면, 방어적인 주식의 베타값은 1 미만   

$e_i$는 잔차를 의미하며 $E(e_i)=0$이기 때문에 양변에 기댓값을 취하면   
$E(r_i) = r_f + \beta_i[E(r_M) - r_f]$  
- 베타가 높은 증권이 시장 위험에 대한 민감도가 커지므로 이러한 위험에 대한 보상으로 더 큰 위험 프리미엄을 제공 → 체계적 리스크 프리미엄  

Markowitz 모델의 문제점 중 하나는 이를 구현하는 데 필요한 파라미터 추정치의 수가 많다는 것. single-index model은 파라미터 추정치의 수가 훨씬 감소함.  
$\frac{n^2+n}{2}$ → (3n + 2)

Total risk = Systematic risk + Firm-specific risk  
$\sigma^2_P = \beta^2_P\sigma^2_M + \sigma^2(e_P)$  
($\beta^2_P\sigma^2_M$ = Systematic risk, $\sigma^2(e_P)$ = Firm-specific risk)  
$Cov(r_i,r_j) = \beta_i\beta_j\sigma^2_m$  
$Corr(r_i,r_j) = Corr(r_i,r_M) * Corr(r_j,r_M)$    

n개 증권으로 구성된 동일한 가중치의 포트폴리오를 선택한다고 가정($w_i = 1/n$)  
$R_p = \alpha_p + \beta_pR_M + e_p$    
$\beta_P = \frac{1}{n}\sum^n_{i=1}\beta_i$  
$\alpha_P = \frac{1}{n}\sum^n_{i=1}\alpha_i$  
$e_P = \frac{1}{n}\sum^n_{i=1}e_i$  
$\sigma^2_P = \beta^2_P\sigma^2_M + \sigma^2(e_P)$  

  
$\sigma^2(e_P) = \frac{1}{n}\bar{\sigma}^2(e)$  
- 다각화로 인해 0으로 수렴하여 $\sigma^2_P = \beta^2_P\sigma^2_M$, 하지만 매우 큰 n의 경우에도 위험의 일부가 남아있기 때문에 systematic risk는 제거할 수 없다.

$\alpha_P = \sum^{n+1}_{i=1}w_i\alpha_i$  
$\beta_P = \sum^{n+1}_{i=1}w_i\beta_i$  
$\sigma^2(e_P) = \sum^{n+1}_{i=1}w_i\sigma^2(e_i)$

## Lesson18. ETF  
ETF: 거래소에서 거래되는 펀드로, 주식, 채권, 원자재, 통화 등 다양한 자산으로 구성된 포트폴리오를 추적하며, 개별 주식처럼 거래소에서 매수 및 매도가 가능  

특정 산업에 대한 포트폴리오의 노출을 줄이기 위해 ETF를 매도하여 헤징할 수 있음  
ex. S&P500 종목을 많이 포함하고 있는 포트폴리오가 있다면 S&P500 ETF를 매도하여 헤징  
  
ETF 스폰서는 AP(금융기관)라고 불리는 특별 파트너하고만 거래를 진행함.  
AP는 주식 시장에서 ETF스폰서가 정의한 포트폴리오와 일치하는 주식을 구매. ETF스폰서는 AP로부터 받은 주식 묶음에 대한 대가로 ETF주식 제공.  
이후 AP는 증권 거래소에서 ETF주식을 판매  
→ 일련의 과정을 Create Process라고도 한다.   

반대로 AP가 주식시장의 투자자들로부터 ETF주식을 매입하고 ETF스폰서에게 ETF주식을 주고 스폰서로부터 원래 주식을 받음. 이후 AP가 받은 주식을 증권 거래소에 판매  
→ Redeem Process라고 한다.  

AP가 이러한 과정을 반복하는 이유는 ETF가치와 해당하는 주식의 가치를 같게 만들어주기 위함.  
1. ETF가격이 주식 가격보다 높아지면 Create Process 진행  
2. ETF가격이 주식 가격보다 낮아지면 Redeem Process 진행   

## Lesson19. Portfolio Risk and Return  
포트폴리오의 위험을 측정하는 척도로 표준편차 활용  
$\begin{align*}
\sigma_P^2 &= \sum_{i} p(i) \left[ r_P - \mathbb{E}(r_P) \right]^2 \\
&= x_A^2 \sigma_A^2 + x_B^2 \sigma_B^2 + 2x_A x_B \text{Cov}(r_A, r_B)
\end{align*}$  
- 두 자산간의 상관관계가 1이라면 완전제곱식 형태가 되면서 포트폴리오 표준편차는 단순하게 개별 종목의 표준편차를 가중평균한 값이 된다.  
- 두 자산간의 상관관계까 -1이라면 $x_A = \frac{\sigma_B}{\sigma_A + \sigma_B}$가 되고 포트폴리오 분산을 0으로 만들어준다.  
→ 현실적으로 체계적 위험을 제거할 수 없기 때문에 불가능  

  
포물선 형태의 위쪽을 Efficient Frontier라고 하며 가장 낮은 변동성 값을 가지는 점에서의 포트폴리오를 최소 분산 포트폴리오라고 한다.  


  
무위험 자산과 시장 포트폴리오를 결합한 새로운 포트폴리오를 구성한다고 가정  
Efficient Frontier와 접하는 선이 CML이 된다.(동일한 위험수준에서 가장 높은 수익)  
이때 shorting이 허용되면 직선을 따라 우측으로 더 나가는 것도 가능

```python
def calculate_cumulative_returns(returns):
    """
    Calculate cumulative returns.

    Parameters
    ----------
    returns : DataFrame
        Returns for each ticker and date

    Returns
    -------
    cumulative_returns : Pandas Series
        Cumulative returns for each date
    """
    
    portfolio_returns = returns.sum(axis=1)
    cumulative_returns = (1 + portfolio_returns).cumprod() 
    
    if returns.iloc[0].isna().all():
        cumulative_returns.iloc[0] = float('nan')
    
    return cumulative_returns

project_tests.test_calculate_cumulative_returns(calculate_cumulative_returns)
```

```python
import cvxpy as cvx

def get_optimal_weights(covariance_returns, index_weights, scale=2.0):
    """
    Find the optimal weights.

    Parameters
    ----------
    covariance_returns : 2 dimensional Ndarray
        The covariance of the returns
    index_weights : Pandas Series
        Index weights for all tickers at a period in time
    scale : int
        The penalty factor for weights the deviate from the index 
    Returns
    -------
    x : 1 dimensional Ndarray
        The solution for x
    """
    assert len(covariance_returns.shape) == 2
    assert len(index_weights.shape) == 1
    assert covariance_returns.shape[0] == covariance_returns.shape[1]  == index_weights.shape[0]

    #TODO: Implement function
    # Number of assets
    n = covariance_returns.shape[0]
    
    # Define the optimization variable
    x = cvx.Variable(n)
    
    # Define the objective function: minimize portfolio variance + penalty for deviation from index weights
    portfolio_variance = cvx.quad_form(x, covariance_returns)
    penalty = scale * cvx.norm(x - index_weights)
    objective = cvx.Minimize(portfolio_variance + penalty)
    
    # Define the constraints: weights must sum to 1
    constraints = [x >= 0, cvx.sum(x) == 1]
    
    # Solve the optimization problem
    problem = cvx.Problem(objective, constraints)
    problem.solve()
    
    # Return the optimized weights as a 1D numpy array
    return x.value


project_tests.test_get_optimal_weights(get_optimal_weights)
```
