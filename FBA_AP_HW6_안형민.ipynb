### 10.1 Multifactor Models: A Preview  

자산 수익률의 불확실성에는 두 가지 원인이 존재  
1. common or macroeconomic factor → 기댓값이 0  
2. firm-specific events  

※ single-factor model  
$R_i = E(R_i) + \beta_iF + e_i$    
$R_i = E(R_i) + \beta_iF + e_i$  
($\beta_P = \sum w_i\beta_i$, $E(R_P) = \sum w_iE(R_i)$)  
$\sigma^2_P = \beta^2_P\sigma^2_F + \sigma^2(e_P)$  
$\sigma^2(e_P) = \sum w^2_i\sigma^2(e_i)$  
→ 체계적 위험을 단일 팩터만으로 설명하는 것에는 어려움이 있다.  

※ multi-factor model  
- GDP성장과 예상하지 못한 이자 변화율 IR 반영  
$R_i = E(R_i) + \beta_{iGDP}GDP + \beta_{iIR}IR + e_i$  

- 두 거시적 팩터는 모두 기댓값이 0이다.  
- GDP 베타는 일반적으로 플러스, 금리 베타는 일반적으로 마이너스가 될 것으로 예상(금리 인상은 기업에게 좋지 못한 소식)  
- 예를 들어 항공사는 경제 활동에는 매우 민감하지만 금리에 대해서는 덜 민감하기 때문에 높은 GDP베타값과 낮은 금리베타 값을 가진다. 경제가 성장하면 GDP와 금리 모두 증가하지만 GDP 베타값이 더 높은 항공사의 경우에는 긍정적인 영향이 생길 것.  


다른 매크로 팩터에 대한 현재 노출을 우려하는 투자자는 동일하지만 반대되는 노출로 포지션을 시작함으로써 해당 위험을 상쇄할 수 있다. → 헤징 전략의 프레임워크 제공  

### 10.2 Arbitrage Pricing Theory  

Stephen Ross로부터 1976년 APT 개발  

Assumption  
1. 증권 수익은 factor model로 설명될 수 있다.  
2. 특이한 위험을 분산시키기에 충분한 증권이 존재한다.  
3. 원활하게 작동하는 증권시장은 차익거래 기회를 허용하지 않는다.  

※ Well-Diversified Portfolios  

$R_P = E(R_P) + \beta_PF + e_P$  → $R_P = E(R_P) + \beta_PF$  

$R_P = \alpha_P + \beta_PR_M$  

  
Y축과 만나는 점의 경우 체계적 요인은 0이다.  
Well-diversified portfolio A는 비체계적 위험이 0인 반면  Single stock B는 비체계적 위험이 존재한다.  

- 실제로는 1,000개 이상의 자산을 보유한 포트폴리오도 흔치 않기 때문에 현실적인 포트폴리오 크기에서는 잔여 위험을 완전히 제거하는 것은 불가능하며, 이는 APT의 한계점이라고 할 수 있다.  

### 10.3 The APT and the CAPM  
1. APT는 CAPM과 동일한 기능을 많이 제공한다.  
→ 자본 예산 편성, 증권 평가 또는 투자 성과 평가에 사용할 수 있는 수익률에 대한 벤치마크를 제공  

2. 둘 다 기업별 위험과 체계적 위험의 차이를 강조한다.  


- CAPM은 모든 개별 자산에 대해 기대 수익률-베타 관계가 성립한다고 주장하는 반면, APT는 일부 개별 자산에서는 이 관계가 성립하지 않을 수 있다고 본다.   

### 10.4 A Multifactor APT   

- APT의 벤치마크 포트폴리오는 팩터 포트폴리오로, 팩터 중 하나에 대해 베타가 1이고 다른 팩터에 대해 베타가 0으로 구성된 잘 분산된 포트폴리오로 설정.  

※ two-factor  
$E(r_Q) = \beta_{P1}E(r_1) + \beta_{P2}E(r_2) + (1-\beta_{P1}-\beta_{P2})r_f$  
- 동일한 베타를 가진 잘 분산된 포트폴리오는 동일한 기대 수익률을 가져야 한다.   

### 10.5 The Fama-French (FF) Three-Factor Model   
※ The Fama-French Three-Factor Model  

$R_{it} = \alpha_i + \beta_{iM}R_{Mt}+\beta_{iSMB}SMB_t + \beta_{iHML}HML_t + e_{it}$  
SMB = Small Minus Big  
HML = High Minus Low  

1. Size  
시가총액 기준으로 소규모 회사가 대형 회사에 비해 평균 수익률이 더 높은 경향이 있음. 

시가총액 기준으로 주식을 정렬한 후   
시가총액 하위 10%는 Small, 상위 10%는 Big으로 설정. 각 그룹의 동일 가중 수익률을 계산한 다음 소형주 수익률에서 대형주 수익률을 뺀 값을 2로 나누어준다.  
→ SMB(Small minus Big)  


2. Value  
시장가치 대비 장부가치(B/M)가 높은 주식이 낮은 주식보다 성과가 좋은 경향이 있다.  
해당 가치가 높은 상위 30%의 주식을 Value(High), 하위 30%의 주식을 Growth(Low)라고 한다.  
Factor return = $\frac{r_{value}-r_{growth}}{2}$   
→ HML(High minus Low)  


3. Market  
기존 CAPM과 동일하게 시장 초과수익률 활용   

※ Fama-French 3 Factor Model 과정  
시가총액 기준으로 정렬하여 Small, Big 선정. B/M기준으로 Value, Neutral, Growth으로 정렬하여 6개의 그룹으로 나누어준다.  
1. Small Value  
2. Small Neutral  
3. Small Growth  
4. Big Value  
5. Big Neutral  
6. Big Growth  

SMB = {(1+2+3) - (4+5+6)} / 3  
HML = {(1+4) - (3+6)} / 2    

이후 5 factor model로 확장되기도 함.  
- RMW (for robust minus weak) → 수익성이 강한 다각화된 주식 포트폴리오와 약한 주식 포트폴리오의 수익률 차이   
- CMA (conservative minus aggressive) → 저투자 기업과 고투자 기업의 주식으로 구성된 다각화된 포트폴리오의 수익률 차이  

※ Smart Betas and Multifactor Models  
- 스마트 베타 ETF를 통해 투자자는 거래하기 쉬운 지수와 같은 상품을 사용하여 다양한 시장 외 위험 요인에 대한 포트폴리오 노출을 조정할 수 있다.

- 각 시간에서의 알파 팩터는 표준화되어 평균이 0이고 절댓값의 합이 1이다.  

Alpha Value: 단일 기간 동안의 알파 벡터의 원소   

하지만 우리가 쉽게 학술 논문에서 볼 수 있는 알파팩터의 성능을 기대할 수는 없다. 이미 시장에 확산됨으로 인해 생산 가치를 잃기 때문

4. 데이터 소스 또는 데이터를 사용하는 새로운 방법에 대해 배울 수 있다.  
 

- Sector neutralization  
알파 팩터의 성능만을 측정하기 위해 리스크 팩터를 통제해주어야 한다.  
→ 수백개 또는 수천개의 주식을 다룰 때 모든 주식의 시장 베타 값을 1으로 가정할 수 있다. 이후 Dollar neutral을 통해 알파값의 합을 0으로 만들어 시장 리스크를 통제한다.  

각 섹터별 알파 값에서 섹터별 평균 알파를 빼서 Sector Risk를 통제한다.  
→ Sector neutral  

- Ranking   
Winsorizing: 95번째 백분위수 초과, 5번째 백분위수 미만의 값들을 각각 95번째 백분위수, 5번째 백분위수 값으로 대체하는 것.    
→ 이상치 처리에 활용   

- Z-scoring    
$Z = \frac{x_i - \mu}{\sigma}$  
일반적으로 활용되는 데이터 표준화 방법 중 하나  
값의 스케일을 조정해줌으로써 데이터의 상대적 위치를 평가할 수 있다.   

- Smoothing    
노이즈가 많은 금융 데이터에 대해 팩터를 더 강건하게 만들어주기 위한 작업  

1. Rolling Window Average    
$\alpha_{t,\text{smoothed}} = \frac{\sum_{t=1}^{T} \alpha_t}{T}$  


2. Weighted Moving Average  
$\alpha_{t,\text{smoothed}} = \frac{(T \times \alpha_t) + ((T-1) \times \alpha_{t-1}) + \dots + (1 \times \alpha_{t-T+1})}{T + (T-1) + \dots + 1}
$  
- 최근 데이터에 더 큰 가중치 부여    


→ 알파 값은 평균이 0, 절댓값의 합이 1이 되도록 표준화되고 이러한 알파 값을 각 자산의 가중치로 활용한다.    

- Look-ahead Bias  
당시에 알려지지 않았던 정보를 사용하여 과거 데이터에 대한 거래를 시뮬레이션할 때 발생하는 오류  
→ ex. Survivorship bias: 현재시점에 살아남은 기업들만을 대상으로 시뮬레이션을 돌리는 과정에서 과거에 파산했거나 인수된 회사는 자동으로 제외되는 경우 발생.  

이를 방지하기 위한 방법으로 상업용 지수 제공자(S&P 500 등)을 활용.  
이는 특정 날짜의 포트폴리오 구성에 대해 정확한 기록을 보관하며, 이는 당시에 포함되었던 모든 회사들을 반영하기 때문에 해당 시점에서 실제로 존재했던 회사들을 모두 포함하여 분석을 할 수 있다.   

- Sharpe ratio  
$S = \frac{r_i-r_f}{\sigma_i}$  
알파 팩터의 상대적 성능을 비교하는데 활용된다.   

- Rank Information coefficient    
알파 값의 순위가 미래 수익률 순위와 상관관계가 있는지의 여부를 알려준다.    

자산의 미래 성과가 알파 팩터에 의한 예상와 일치하면 정보 계수가 더 높아진다.  

Foward return: 알파 값이 계산된 시점과 비교하여 수익이 미래에 발생할 수 있는 수익률  
→ t시점 이전에 알파 값이 계산되면 t시점부터 t+1시점까지의 수익률을 계산  

순위 값의 상관관계를 파악하기 위해 스피어만 상관계수 활용. 공식은 피어슨 상관계수와 동일하나 변수 X, Y가 순위로 변환된다.  
→ t-1시점의 알파 팩터 값과 t~t+1시점의 수익률 간의 상관관계 계산  


- Information ratio  

return = systematic return + specific return  
$r = \sum_{k=1}^{K} (\beta_k f_k) + s$  
→ IR은 specific return에 적용되는 샤프비율이다.  

이를 통해 펀드매니저가 포트폴리오에 기여하는 성과를 측정하는 샤프비율로 해석할 수 있다.  

- Fundamental Law of Active Management   
$IR = IC \times \sqrt{B}$  
IR: Information Ratio  
IC: Information Coefficient  
B: Breadth(투자 결정의 빈도나 독립적인 의사결정 기회의 수)   


- 샤프비율을 개선하는 방법  
1. 알파 팩터가 올바른 주식을 정확하게 선택할 수 있는 확률을 높이는 것(IC)  
2. 독립적인 베팅 횟수를 늘리는 것(breadth)   


※ Real World Constraints

2. Transaction costs  


- Turnover analysis  
  
포트폴리오의 총 가치 중 일정 기간 동안 거래되는 비율   

회전율이 높으면 그만큼 거래 빈도가 많다는 것을 의미한다. 거래 비용(수수료, 스프레드, 세금 등)은 거래 횟수가 많을수록 증가하므로, Turnover 분석을 통해 거래 비용이 포트폴리오에 미치는 영향을 평가할 수 있다.  


- Quantile analysis    
알파 값을 지정한 분위수로 나누고 각 그룹의 평균 수익률과 표준편차를 계산하여 비교하는 것.   

학술 연구의 관점 (Academic Research): 넓게 적용 가능한 시장 현상을 발견하는 것이 주된 목표이기 때문에 꼬리에 중점을 둔다.  

실무자의 관점 (Practitioner): 개별 자산(포트폴리오 내의 각 주식)에 대해 구체적인 거래 결정을 내리는 것이기 때문에 모든 분위수 그룹에 관심을 갖는다.  


- Transfer Coefficient  
$corr(\alpha, x)$  
최적화된 포트폴리오 가중치가 원래 알파 벡터와 얼마나 일치하는지 측정하기 위해 사용  
→ 높은 Transfer Coefficient는 알파 벡터가 포트폴리오 최적화 도구의 위험 중립화에서 살아남았다는 좋은 신호이다.  

- Conditional factors  
어떤 자산에 대해 팩터 A에 대한 알파 값과 팩터 B에 대한 알파 값을 곱하여 조건화된 알파 값을 생성할 수 있다.

따라서 주간 수익률이 높은 주식에 비중을 늘리고, 주간 수익률이 낮은 주식에 비중을 줄이는 전략을 세울 수 있음.  

2. "The Formation Process of Winners and Losers in Momentum Investing"  

  
어떤 기간까지의 두 주식의 수익률이 동일하더라도 그 과정이 다르다면 어떤 주식을 선호해야 하는가?  
→ 주식의 궤적이 모멘텀의 가속 또는 감속 여부를 나타내는 지표가 되는 방식에 대해 논의  
   
convex한 형태를 띄는 그래프가 concave한 그래프보다 미래 수익률이 더 높을 수 있다. 
따라서 Accelerated Gains를 많이 매수하고 Accelerated Loss를 많이 매도하는 것이 좋다.  
  
→ 하지만 상대적인 convex / concave 에 대한 기준이 모호하다는 문제가 있다.   


독립변수가 시간 t, $t^2$이고 종속변수가 주가인 다중회귀 분석 $y = Gain * t + Acc * t^2$ 으로 설정.  
Gain의 부호가 동일할 때 Acc가 양수인지 음수인지에 따라 Accelerated or Decelerated 결정  
   

Ex. Stock A: Gain 10, Acc 2    Stock B: Gain 10, Acc -2  
→ A는 Accelerated Gains이고 B는 Decelerated Gains 이므로 A에 더 많이 매수   

- Gain, Acc 계수를 순위로 변환한 다음 두 값을 곱하면 알파 팩터를 만들 수 있다.   

3. "Expected Skewness and Momemtum"  

Attentional Bias: 사람들이 큰 그림에 초점을 맞추는 것이 아닌 특정 측면에 집중하고 이전의 기존 신념을 뒷받침하는 증거에 더 주목하는 현상   

뉴스에 더 많이 보도되는 주식의 일일 수익률이 높았다면 개인 투자자들은 이를 주식이 계속 상승할 것이라는 신호로 여길 수 있다.  
→ 계속 주가가 상승하다가 어느순간 주식이 과도하게 매매되고 있음을 깨닫고 주가가 다시 바닥으로 떨어지게 된다.    

왜도(Skewness): 분포의 치우침을 나타내는 척도   
  

첨도(Kurtosis): 분포의 꼬리가 얼마나 두껍거나 얇은지를 설명하는 척도  
- 주식 수익률은 일반적으로 정규분포에 비해 높은 첨도를 나타낸다.  

다음달의 일일 최대 수익률로 skewness를 확인하고 지난해의 수익률이 모멘텀 벡터라고 가정  

1. 모멘텀 양수, 왜도 양수  or  모멘텀 음수, 왜도 음수
→ Weakened momentum  

2. 모멘텀 양수, 왜도 음수  or 모멘텀 음수, 왜도 양수  
→ Enhanced momentum  


모멘텀과 왜도를 모두 순위로 변환   
→ joint factor: Momentum * Skew    


4. "Arbitrage asymmetry and the idiosyncratic volatility puzzle"  

Ivol(Idiosyncratic Volatility): 개별 자산이나 기업의 고유한 변동성   
ivol은 차익거래 위험을 나타내는 좋은 지표가 될 수 있다.  
→ ivol 자체가 conditioning factor이다. 

Fundamental factor: 기업의 재무 상태, 성장성, 수익성, 가치 평가 등을 측정하는 다양한 지표로 구성   
→ Fundamental factor와 ivol을 결합하여 conditional factor 생성 가능  

- conditional information은 그 자체로 알파 팩터가 될 필요가 없다.
