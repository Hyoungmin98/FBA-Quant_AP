# Chapter 6. Fundamental Factor Models  
## 6.1 INTRODUCTION  
- 팩터 익스포져는 주식이 어떤 종류의 위험에 노출되어 있는지를 나타낸다.  
- 팩터 프리미엄은 그 위험을 감수함으로써 투자자들에게 주어지는 보상을 정량화함  

평균 주식 수익률 = 팩터 익스포져 * 팩터 프리미엄  

- 펀더멘털 팩터 모델에서 팩너 익스포져는 알 수 있음  
→ 주식의 시가총액이나 PBR과 같은 주식의 관찰 가능한 펀더멘털 특성  

- 반면 팩터 프리미엄은 알 수 없기 때문에 추정해야 한다.  

$r_i = \alpha + \beta_{i1}f_1 + \ldots + \beta_{iK}f_K + \varepsilon_i$  
오차항의 평균이 0이므로 주식의 기대수익률은 다음과 같다.  
$E(r_i) = E(\alpha) + \beta_{i1}E(f_1) + \ldots + \beta_{iK}E(f_K)$   

여기서 리스크는 분산 가능한 리스크, 분산 불가능한 리스크 두가지로 나뉨  

$\text{총 위험 = 분산 불가능한 위험 + 분산 가능한 위험}$    
$V(r_i) = V(\alpha + \beta_{i1}f_1 + \ldots + \beta_{iK}f_K) + V(\varepsilon_i) \\
= V(\beta_{i1}f_1 + \ldots + \beta_{iK}f_K) + V(\varepsilon_i)$  

- 펀더멘털 팩터 모델은 주식의 수익률과 위험을 예측하는 데 사용될 수 있으며, 이러한 예측은 포트폴리오 매니저가 자신의 포트폴리오에 가장 적합한 주식을 선택하는 데 도움을 줄 수 있다.  

1. 팩터 익스포져를 관찰하고 팩터 프리미엄을 추정하기 전에 모델에 포함할 팩터들을 결정하는 것, 무위험 이자율의 처리, 데이터의 시간 간격과 기간, 투자 유니버스 등 결정  
2. 개별 주식의 팩터 익스포져 값 구하기  
3. 팩터 프리미엄과 상수 alpha 추정  
4. 분산 가능위험과 분산불가능 위험을 추정함으로써 총 위험에 대해 평가  

※ 구체적인 과정  

- 예비작업  
1. 모델에 포함할 팩터 선택  
2. 무위험 이자율 처리 결정  
3. 투자 유니버스 정의  
4. 데이터의 시간 간격과 기간 결정  

- 주식 총수익률 모델링  
1. 다중회귀분석으로 주식 수익률 추정  
2. 각 시간 간격에서 기간에 대한 주식수익률과 팩터 익스포져 데이터 수집  
3. 다중회귀분석을 통해 팩터 프리미엄 추정  
4. 추정값의 견고함을 확인하고 각 부분 집합에 대한 추정치 비교. 추정치가 비슷하면 추정이 견고함  
5. 추정치가 견고하지 않다면 다른 추정 방법 시도  
6. 각 주식의 평균 수익률 계산  
7. 주식 수익률의 위험을 분산 가능, 분산 불가능 위험으로 분류  
8. 투자 유니버스 내의 주식 수익률 간 상관관계 계산  


- 벤치마크 매니저를 위한 잔여 주식 수익률 모델링  
1. $\tilde{r}_i = \tilde{\alpha}_i + \tilde{\beta}_i \tilde{r}_B + \tilde{\varepsilon}_i$  
2. 주어진 기간의 주식수익률과 벤치마크 수익률 데이터 수집  
3. $\tilde{\alpha}$와 $\tilde{\beta}_i$를 추정한 후 잔여수익률 계산: $
\tilde{r}_i ≡ \tilde{\alpha}_i + \tilde{\varepsilon}_i$    
4. 다음 방정식으로 잔여 수익률을 추정: $\tilde{r}_i = \alpha + \beta_{i1}f_1 + \ldots + {\beta}_{iK}f_K + {\varepsilon}_i$  
5. 주식 총수익률 모델링을 위한 주요 단계 2~8을 따르되 총수익률 대신 잔여 수익률로 대체    


## 6.2 PRELIMINARY WORK  
### 6.2.1 Choosing Factors  
- 팩터들은 위험을 나타냄  
- 여러 종류의 팩터가 있지만 이 중 일부는 주식 특성을 설명하고 일부는 전체 시장의 상황을 설명  
- 펀더멘털 팩터는 주식 자체의 특성이며, 일반적으로 재무제표에서 얻을 수 있는 값  
→ 일부 기술적 팩터와 애널리스트 팩터도 펀더멘털 팩터로 간주  

※ 대표적인 펀더멘털 팩터 리스트  
1. **펀더멘털 밸류에이션 팩터: 배당 수익률, 장부가 대비 가격 비율, 수익 대비 가격 비율, 매출 대비 가격 비율**
2. **펀더멘털 사이즈 팩터: 시가총액의 로그값**
3. **펀더멘털  지급능력 및 재무위험 팩터: 부채 대비 자본 비율, 유동비율, 역이자 커버리지 비율**
4. **펀더멘털 운영 수익성 팩터: 순이익 성장, 자산 대비 수익률, 보통주 자본 대비 수익률**
5. **기술적 팩터: 12개월 모멘텀, 거래량, 공매도 비율**
6. **애널리스트 팩터: 애널리스트 등급 변화 및 어닝 리비젼**    

- GDP와 같은 경제적 팩터나 소비자 심리와 같은 행동적 팩터, 는 펀더멘털 팩터 모델에 맞지 않는다.  

- 일부 실무자들은 P/E를 사용하고 손실이 있는 주식을 제외하는 반면, 다른 이들은 E/P를 사용하고 선택할 수 있는 주식의 범위를 더 넓게함  
→ 핵심은 일부 흔한 팩터들의 역을 사용함으로써, 우리는 훨씬 더 많은 주식을 분석할 수 있으며, 이는 주식 선택에 대한 훨씬 풍부한 관점을 제공한다는 것  

### 6.2.2 Treatment of the Risk-Free Rate  

위험이 실제로 보상되는 주식 수익률 부분에 초점을 맞추기 위해, 일부 포트폴리오 매니저들은 펀더멘털 팩터 모델을 구현하기 전에 주식 수익률에서 무위험 이자율을 차감함  
→ 이론적으로 타당하지만, 실제로 이를 수행하기는 쉽지 않다. 현실에서 완벽한 무위험 자산이란 존재하지 않기 때문. 그럼에도 불구하고 미국 재무부 증권을 무위험 자산으로 간주  

### 6.2.3 Choosing the Time Interval and Time Period  
데이터 포인트 간에 어느 정도의 시간 간격이 있어야 하며, 데이터 세트가 커버해야 할 전체 기간은 얼마나 되어야 할까?  
- 이상적으로, 시간 간격은 투자 기간(즉, 리밸런싱 빈도)을 반영해야 함  
→ 그러나 추정 정밀도를 고려하면 포트폴리오 매니저는 투자 기간과 다른 시간 간격을 선택해야 할 수도 있음  

- 그럼에도 불구하고, 긴 시간 간격보다는 짧은 시간 간격에 오류를 범하는 것이 일반적으로 가장 좋음  
- 실무에서 금융 경제학자들은 월간 간격을 선호하며, 일부 매니저는 주간 간격을 선호한다.  
- 시간 간격이 짧을수록 수집 데이터 양은 많아지게 된다.  

- 금융 경제학자들은 월간 간격을 사용할 때 분석에 36에서 60의 시간 간격을 포함하는 경향이 있음(총 기간은 3~5년)

### 6.2.4 Choosing the Universe of Stocks  
- 포트폴리오의 투자 유니버스에 대한 외부 제한이 이미 있을 수 있음  
→ 포트폴리오 매니저는 잠재적 투자 목록을 좁힐 수 있는 자체적인 초기 스크리닝 전략을 가질 수 있음   
- 투자 유니버스에 대한 외부 제한이 없다면 유니버스를 굳이 축소시킬 필요는 없음

→ 주식 선택 풀이 클수록 풀 어딘가에 수익률이 높은 주식이 있을 가능성이 더 높지만 그만큼 주식 상관관계 계산의 정확도가 떨어지며 모델의 신뢰도가 떨어짐

## 6.3 BENCHMARK AND α  

많은 포트폴리오 매니저들은 자신들의 포트폴리오 성과를 측정하는 데 특정 벤치마크를 사용한다.  

- 펀더멘털 팩터 모델에 벤치마크를 통합하는 한 가지 접근 방식은 모델을 사용하여 전체 주식 수익률이 아니라 잔여 수익률만을 예측하는 것  

$r_i=\tilde{\alpha_i}+\tilde{\beta_i}r_B+\tilde{\epsilon_i}$  
→ 이를 통해 $\tilde{\alpha}$와 $\tilde{\beta}$를 추정한 후, 잔여 수익 $\tilde{r_i}$를 다음과 같이 정의한다.  
$\tilde{r}_i = \tilde{\alpha}_i + \tilde{\epsilon}_i$    

이후 포트폴리오 매니저는 펀더멘털 팩터 모델을 사용하여 잔여 수익률의 가치를 예측할 수 있다.  

- 벤치마크를 고려하는 또 다른 방법은, 잔여 주식률을 예측하는 대신에, 벤치마크를 모델에 하나의 팩터로 추가하는 것  
→ 이 접근 방식은 경제적 팩터 모델과 더 잘 어울림  


## 6.4 FACTOR EXPOSURE   
팩터 노출은 주식의 위험에 대한 노출을 정량화할 수 있다.  
→ 펀더멘털 팩터 모델에서 팩터 노출은 주식의 관찰 가능하거나 쉽게 계산 가능한 특성의 값  

팩터 노출은 시간이 지남에 따라 변하고, 우리는 올바른 날짜에 올바른 값을 할당해야 함  
→ 월간 간격의 경우, 월 초에 기록된 팩터 노출을 사용하는 것이 표준 관행  


회사의 재무제표에서 데이터를 적용하는 데에는 특별한 주의가 필요  
- 회사들은 분기별로 재무제표를 발표하지만, 회계 분기가 끝난 직후에 즉시 발표되는 것은 아님. 일반적으로 회계 기간이 끝나고 변수를 보고하는 데에는 2~3개월의 지연을 허용하는 것이 합리적  → 백테스팅 과정에서 사후편향 조심  

## 6.5 THE FACTOR PREMIUM  
팩터 프리미엄은 주식의 팩터 익스포져의 각 단위에 대한 보상  

회귀분석을 통한 팩터 프리미엄의 추정은 프리미엄이 여러 해 동안, 그리고 주식 간에 일정하게 유지될 가능성이 있기 때문에 가능  
$r_{it} = \alpha + \beta^{'}_{it}f_t + \varepsilon_{it}$  
가장 간단한 방법은 OLS 접근법.  
→ 구하기 쉽지만 가장 신뢰할 수 있는 추정치는 아닐 수 있다.  


### 6.5.1 OLS Estimator of the Factor Premium  

    
$\hat{\alpha}
 = \bar{r}-\hat{\beta^{'}}\hat{f}$    
 
  
$\hat{\sigma}^2$ 은 잔차의 분산  

   

- 표준오차는 추정치의 정밀도를 나타냄  
→ 작은 표준오차는 추정치가 매우 정밀하다는 것을 의미  

  
위의 표는 팩터 프리미엄 추정치와 표준오차를 나타낸 표  
→ E/P 팩터 프리미엄의 진정한 값은 대략 −4.235와 −2.615 사이라고 95% 확신할 수 있고, M12M 프리미엄의 진정한 값은 대략 −2.807와 −1.999 사이라고 말할 수 있음(Confidence Interval)  
- 이 표에서는 E/P 팩터 프리미엄이 음수로 나와 기존의 흐름(?)과는 역으로 나타났다는 것을 알 수 있다. 하지만 단기적으로 실현된 프리미엄의 경우 장기적인 프리미엄과는 다를 수 있으며 실제로 2016~2020년까지 가치팩터의 실패에 대한 많은 논의가 있었다  


## 6.5.2 Robustness Check   
모델은 보통 현실을 정확하게 묘사하지는 않고, 이에 대한 좋은 근사치일 뿐이다.  

- 통계학에서 모델의 부족한 점을 설정 오류라고 부른다.  
- 설정 오류는 모든 회귀분석에서 발생. 따라서 지속적이고 안정적인 패턴을 반영하는 모델을 구축하기 위해 노력해야한다.   

강건성 검사를 통해 팩터 프리미엄 추정치가 안정적인지를 판단할 수 있다.  
→ 추정치가 강건하지 않다면 다른 추정 기법을 시도해야함  
- 강건성을 확인하기 위해 데이터 세트를 하위 집합으로 쪼개서 분석  
ex. 2016년 1월부터 2018년 6월까지의 기간과 2018년 7월부터 2020년 12월까지의 기간에 대한 팩터 프리미엄 추정, 또는 섹터별로 하위 집합을 구성하여 팩터 프리미엄 추정  
→ 추정이 강건하다면 추정치가 너무 다르지 않아야한다.  


### 6.5.3 Outliers and MAD Estimator of Factor Premium  
OLS 방법은 이상치에 너무 민감하다는 단점 존재  
- 강건성 검사를 통해 OLS 추정치가 매우 불안정하다는 것이 확인된다면 이상치에 덜 민감한 다른 추정방법을 사용할 필요가 있음. 그중 최소 절대 편차(MAD) or 중앙값 추정이 대안이 될 수 있다.  

- MAD는 제곱 잔차가 아닌 잔차의 절대값 합을 최소화  
→ 제곱을 하지 않기 때문에 이상치에 대한 영향이 적음  

### 6.5.4 Heteroscedasticity and Autocorrelation Consistent Estimation of the Standard Error   

- OLS추정에 단점이 존재하긴 하지만 여러 세대에 걸쳐서 많은 분석가들이 활용해왔음  
- 또한, 다른 추정 기법들 또한 그 기법만의 가정이 존재하기 때문에 결국 단점은 존재하기 마련  
→ 이러한 점을 고려할 때 널리 사용되는 한 가지 접근 방식은 OLS 추정치를 계속 보고하면서 동시에 OLS 추정치의 특정 가정을 완화하도록 표준 오류를 조정하는 것  

1. 서로 다른 종목의 모든 오차항이 동일한 분산을 갖는다.(등분산성)  
2. 서로 다른 기간의 오차항이 독립적이다.(독립성)  
→ 이 두 가지 가정을 완화한 후 얻은 표준 오차를 이분산성 및 자기상관 일관성 표준 오차 또는 간단히 HAC 표준 오차라고 한다.  

HAC를 사용하면 서로 다른 회사의 오류 항이 상호 연관되는 것을 허용하지 않는다.(다른 기간에 대한 한 회사의 오류 항이 상호 연관될 수 있음). 또한 HAC는 서로 다른 기간에 대한 한 회사의 오류 항이 동일한 분산을 갖는다(단, 다른 회사의 오류 항이 다른 분산을 가질 수 있음)   
→ 이렇게 완화한 OLS 추정치는 효율적이지 않을 수 있지만 표본이 충분히 큰 경우에는 정확한 편    

## 6.6 DECOMPOSITION OF RISK  
최적 포트폴리오를 구축하기 위해서는 주식 수익률의 위험을 알아야 함  

- 펀더멘털 팩터 모델에서 주식 수익률의 위험은 두 부분으로 구성됨
    - 팩터 프리미엄의 무작위성에서 비롯된 분산 불가능한 위험 $\beta_{i}^{'}\hat{f}$
    - 분산 가능한 위험 $\hat{\epsilon}_i$
    - 주식 수익률의 총 위험은 단순히 이 두 위험의 합  

표본 분산(분산 불가능한 위험. 매월 t월에 대한 회귀분석을 진행하여 얻은 시계열 추정치 활용): $
\frac{1}{T} \sum_{t=1}^{T} \left[ \hat{\alpha}_t + \beta^{'}_{it} \hat{f}_t -\frac{1}{T} \sum_{t'=1}^{T} \left( \hat{\alpha}_{t'} + \beta^{'}_{it'} \hat{f}_{t'} \right) \right]^2$  

분산 가능한 위험: $
\hat{\sigma}_i^2 = \frac{1}{T} \sum_{t=1}^{T} \left( r_{it} - \hat{\alpha}_t - \hat{\beta}^{'}_{it}\hat{f}_t \right)^2$    



- 최적 포트폴리오를 찾는 단계에 도달했을 때, 주식 수익률 간의 상관관계를 알아야 한다.  
- 주식의 총 수익률은 분산 불가능한 부분과 분산 가능한 부분으로 구성  
$
C(r_{it}, r_{jt}) = C(\hat{\beta}_{it}f_t, \hat{\beta}_{jt}f_t) + C(\hat{\epsilon}_{it}, \hat{\epsilon}_{jt})$  
→ 실무적으로는 두 주식 수익률의 분산 불가능한 부분의 상관관계만 추정  

1. N개의 주식이 있다면 N(N-1)/2 개의 공분산을 추정해야 함  
2. 분산 가능한 수익률이 주식 수익률에 거의 영향을 미치지 않음  
→ 관례적으로 분산 가능한 위험 $C(\hat{\epsilon}_{it}, \hat{\epsilon}_{jt})$은 0 이라고 가정

## 7.1 INTRODUCTION   

경제적 팩터모델 또한 주식 수익률이 위험을 감수하는 대가라는 중심 아이디어를 여전히 표현한다.  
→ 모델의 구조도 펀더멘털 팩터 모델과 동일  

- 그러나 경제적 팩터 모델에서는 팩터 노출과 팩터 프리미엄의 역할이 반전됨    
- - **펀더멘털 팩터 모델**에서 팩터 노출은 재무제표에서 관찰할 수 있는 반면, 팩터 프리미엄은 횡단면 회귀에서 추정해야 함  
- **경제적 팩터 모델**에서 팩터 프리미엄은 알려진 값(또는 주어진 데이터에서 계산할 수 있는 값)이며, 팩터 노출은 주식 수익률을 팩터 프리미엄에 대한 회귀로 추정해야 함  

- 경제적 팩터모델은 모든 주식에 영향을 미치는 시장 외부의 위험을 나타내는 경제적 팩터에 특히 의미가 있다.  

ex. 인플레이션율   
팩터 프리미엄은 인플레이션에 영향을 받는 주식에 투자할 때 요구하는 보상의 양  
→ 하지만 이 값은 실제 인플레이션율과 다를 가능성이 크다.  
1. 인플레이션율은 리스크 측면에서 표현되지 않으므로 1% 인플레이션율에 대한 1% 예상 수익률이라고 말할 수 없음  
2. 모든 인플레이션은 위험에 해당하지 않음  

그럼에도 불구하고 경제적 팩터 모델은 선형회귀로 이루어진 모델이기에 경제적 팩터 모델은 투자자들이 요구하는 보상의 양이 실제 인플레이션율과 동일하다고 가정하지 않음. 따라서 인플레이션율을 팩터 프리미엄으로 사용한다.  

$
r_i = \alpha_i + \beta_{i1}f_1 + \ldots + \beta_{iK}f_K + \varepsilon_i$  
- $f_1, \dots ,f_K$는 팩터 프리미엄  
- $\beta_{i1},\dots,\beta_{iK}$는 팩터 익스포져  
- $\alpha_i$는 상수  
- $\beta_{i1}f_1 + \dots + \beta_{iK}f_{1K}$는 주식의 분산 불가능한 위험, 오차 $\epsilon_i$는 주식의 분산 가능한 위험  

- 학계에서는 경제적 팩터 모델이 오랫동안 유일하게 유효한 것으로 간주되어 왔으며, 펀더멘털 팩터 모델은 최근에야 일부 신뢰를 얻고 있다.  
- 학술 문헌에서는 경제적 팩터 모델 or 펀더멘털 팩터 모델이라고 구분하지 않고 멀티팩터 가격 모델이라는 용어를 사용하여 경제적 팩터 모델의 틀을 일반적으로 사용  

- 초기작업은 펀더멘털 팩터 모델과 동일  

- 주식 총 수익률 모델링을 위한 주요 단계(벤치마크 매니저가 아닌 경우)  
1. 주식 수익률은 다음 공식으로 추정됨: $r_i = \alpha_i + \beta_{i1}f_1 + \ldots + \beta_{iK}f_K + \varepsilon_i$  
2. 각 시간 간격에 대해 팩터 프리미엄 데이터 수집    
2.1.  팩터 프리미엄 데이터는 경제/행동/시장 팩터로 구한다.  
2.2. 펀더멘털/기술적/애널리스트 팩터에 대한 팩터 프리미엄을 계산하기 위해서는 제로 투자 포트폴리오 방법 사용    
2.3. 통계적 팩터의 팩터 프리미엄을 계산하기 위해서는 PCA 방법 사용    
3. 이후 시계열 회귀를 사용하여 주식 수익률에 대한 팩터 익스포져 추정   
4. 충분한 팩터 프리미엄 데이터가 없을 경우 유사한 기업을 proxy 사용할 수 있는 기업 특성 매칭을 사용하여 데이터가 부족한 기업 식별  
4.1. proxy의 주식 수익률과 팩터 프리미엄 데이터 수집  
4.2. proxy 데이터를 사용하여 주식 수익률에 대한 팩터 익스포져를 추정  
5. 각 주식의 평균 수익률을 팩터 익스포져 * 팩터 프리미엄으로 계산  
6. 주식 수익률의 위험을 분산 가능한 위험 + 분산 불가능한 위험으로 분해  
7. 투자 대상 주식의 수익률 간 상관관계 계산

## 7.3 BENCHMARK AND α  

- 많은 퀀트 포트폴리오 매니저들은 벤치마크를 기준으로 포트폴리오를 관리한다.  
- 모델을 만들기 시작할 때 벤치마크를 고려하는 것은 유용하지만 필수적이지는 않음.  

- 포트폴리오 성과 측정에서, 벤치마크의 영향을 반영하거나 조정하는 모델을 생성하는 것이 유용할 수 있다.  
→ 모델을 구축하기 전에 주식 수익률에서 벤치마크 관련 수익률을 제거  
1. 주식 수익률을 벤치마크 수익률에 회귀시키고 이를 통해 얻은 잔차를 계산   
2. 총 주식 수익률이 아닌 잔차 주식 수익률에 대한 경제적 팩터모델 구축  
3. 결과적으로 나온 모델은 팩터 프리미엄, 팩터 익스포져, 벤치마크 수익률을 넘어서는 주식 수익률 부분 간의 관계를 보여준다.  

- 대안으로 경제적 팩터모델에 명시적으로 벤치마크를 포함할 수도 있음  

- 벤치마크 수익률을 $r_B$라고 가정  
→ $r_i = \alpha_i + \beta_{i,1}f_1 + \ldots + \beta_{i,K-1}f_{K-1} + \beta_{i,K}r_B + \epsilon_i$  
$
\alpha^B = \alpha_i + \beta_{i,1}f_1 + \ldots + \beta_{i,K-1}f_{K-1}$    
벤치마크 알파는 포트폴리오 매니저와 모델이 주식 수익률에 기여하는 부분을 보여주며, 포트폴리오 매니저의 성과를 측정하는 방법  


## 7.4 THE FACTOR PREMIUM  
경제적 팩터 모델을 생성하기 위해서 포트폴리오 매니저는 주식 수익률과 팩터 프리미엄의 쌍을 수집해야 한다.  

- 경제적 팩터 모델에서, 팩터 프리미엄은 알려진 값.  
→ 하지만 이것이 항상 팩터 프리미엄을 직접 관찰할 수 있다는 의미는 아니다.  

### 7.4.1 Factor Premium for Economic/Behavioral/Market Factors  
- 경제적/행동적/시장 팩터에 대한 팩터 프리미엄을 얻는 것은 어떠한 계산도 필요하지 않음  

  


### 7.4.2 Factor Premium for Fundamental/Technical/Analyst Factors 
  
- 펀더멘털/기술적/애널리스트 팩터에 대한 팩터 프리미엄을 찾기 위해서는 더 많은 계산이 필요함  
→ 제로 투자 포트폴리오를 구성하고 그 수익률을 계산하는 작업이 포함  

- 일반적으로 십분위 or 오분위 or ...로 주식들을 묶고 상위그룹 롱, 하위그룹 숏 하여 구성  
- 제로 투자 포트폴리오 구성의 빈도는 정렬 변수가 자주 업데이트 되지 않는 경우 신중하게 선택되어야 함  
ex. 장부가-가격 비율의 경우 장부가는 분기마다, 가격은 매일 업데이트 되기 때문에 장부가가 업데이트 될 때 비율이 큰 변동을 할 가능성이 크다.  

펀더멘털/기술적/애널리스트 팩터에 대한 팩터 프리미엄 계산 단계  
1. 시간 1에서 모든 주식을 팩터에 따라 순위화  
2. 리스트의 상위 그룹과 하위 그룹을 동일가중하여 고노출 포트폴리오와 저노출 포트폴리오를 만든다.  
3. 고노출 포트폴리오와 저노출 포트폴리오의 수익률 차이로 제로 투자 포트폴리오 수익률 계산. 이 값은 시간 1에서의 팩터 프리미엄이 된다.  
4. 시간 2에 대해 이러한 단계들 반복  
→ 데이터 부족으로 인해 시간 2에 대한 새로운 포트폴리오 구성이 불가능한 경우, 시간 1에 사용된 동일한 포트폴리오에 대해 시간 2의 수익률을 계산  
5. 각 팩터에 대해 각 시간 간격마다 위 절차들 반복  


### 7.4.3 Factor Premium for Statistical Factors  
통계적 팩터에 대한 팩터 프리미엄을 얻기 위해서는 다소 집약적인 계산이 필요하다.  

- 이는 주성분분석(PCA)로 알려져 있으며 시작은 주식 수익률의 분산-공분산 행렬을 추정하는 것  
  

- 이후 직교행렬 Q를 찾아 대각화 진행  
  
- 여기서 D는 대각 행렬이며, 대각 요소는 의 고유값이다. Q의 각 열은 고유값에 해당하는 고유벡터로 이루어진 직교행렬

  
이러한 팩터 프리미엄 값들은 주식 수익률의 선형 조합으로 구성된 독립변수 집합 중에서 N개의 주식 수익률에 대한 가장 높은 표본 내 설명력을 가진다.   


## 7.5 FACTOR EXPOSURE  
### 7.5.1 The Standard Approach  
팩터 익스포져는 때때로 팩터 민감도나 팩터 로딩이라고도 불린다.  
- 계수 $\beta_i$는 우리가 발견하고자 하는 팩터 익스포져이며, $\epsilon_{it}$는 주식 수익률의 분산 가능한 위험을 반영하는 오차 항  

$
\hat{\beta}_i = \left[ \sum_{t=1}^{T} (f_t - \bar{f})(f_t - \bar{f})' \right]^{-1} \left[ \sum_{t=1}^{T} (f_t - \bar{f})(r_{it} - \bar{r}_i) \right]$    

$ 
\hat{V}(\hat{\beta}_i) = \sigma_{i}^{2} \left[ \sum_{t=1}^{T} (f_t - \bar{f})(f_t - \bar{f})' \right]^{-1}$  

$ 
\hat{\sigma}_{i}^{2} = \frac{1}{T} \sum_{t=1}^{T} (r_{it} - \hat{\alpha}_i - \hat{\beta}_{i,t} f_t)^2$  

  
다음 표는 다섯 가지 팩터 모델을 위한 선정된 주식들의 팩터 익스포져 추정치와 표준오차  

- 실업률 팩터 익스포져 값은 MSFT, JNJ, AMZN은 -, AAPL, WMT는 +  
- 이 점은 경제 사이클 방어주로써 애플가 월마트의 위치를 반영한다. 경제가 느려질 때 가치가 상승하는 경향이 있는 주식들  

- 사이즈 팩터의 경우 AAPL, AMZN은 -값을, 나머지는 +값의 팩터 익스포져를 가지고 있음  

- B/P 팩터의 경우 월마트는 가치주, 다른 주식들은 성장주의 성향을 띄고 있음을 알 수 있다.  


$r_i = \alpha_i + \beta_i^{'}f+\epsilon_i$ 에서 팩터 익스포져 추정치는 상식으로 설명될 수 있지만, 이것이 반드시 펀더멘털 팩터 모델에 사용한 팩터 익스포져와 동일하다는 것을 의미하지는 않는다.  

###  7.5.2 When the Standard Approach Fails  
두 회사가 합병할 때 자연스럽게 해야 할 일은 합병 전 두 회사의 팩터 익스포져에 대한 가중 평균을 찾는 것  
→ 적절한 가중치는 합병 전 회사의 시가 총액  
   

## 7.6 DECOMPOSITION OF RISK  
### 7.6.1 The Standard Approach   
주식 수익률의 전체 위험은 분산 불가능한 위험과 분산 가능한 위험의 합  
$ 
V(r_i)=\beta_i^{'}V(f)\beta_i+V(\epsilon_i)$  

$
\hat{V}(f) = \frac{1}{T} \sum_{t=1}^{T} (f_t - \bar{f})(f_t - \bar{f})'$  

- 최적 포트폴리오를 찾는 단계에 이르렀을 때, 우리는 주식 수익률 간의 상관관계도 알아야 함  
→ 두 주식 수익률 간의 상관관계는 분산 불가능한 구성요소 간의 상관관계와 분산 가능한 구성요소 간의 상관관계 두 부분으로 나뉨  

$
C(r_i, r_j) = \beta_i'V(f)\beta_j + C(\epsilon_i, \epsilon_j)$  

- 하지만 실제로는 $T$가 상당히 크지 않는 한, $C(\epsilon_i, \epsilon_j)$는 0이라고 가정하는 것이 일반적이다.  
- N개의 주식이 있다면 $N(N-1)/2$개의 공분산 추정치가 필요  
- 모델이 좋다면 공분산을 0이라고 가정하는 것이 너무 큰 왜곡을 불러일으키진 않는다.  
- 분산 가능한 위험($\epsilon$ 항)들은 포트폴리오에서 분산화되어 무시될 수 있으므로, 그들의 공분산도 무시할 만큼 미미해야 함
