## Chapter9. The Capital Asset Pricing Model   

- Assumption  

1. Individual behavior  
a. Investors are rational, mean-variance optimizers.  
b. Their common planning horizon is a single period.  
c. Investors all use identical input lists, an assumption often termed homogeneous expectations. Homogeneous expectations are consistent with the assumption that all relevant information is publicly available.  

2. Market structure  
a. All assets are publicly held and trade on public exchanges.   
b. Investors can borrow or lend at a common risk-free rate, and they can take short positions on traded securities.  
c. No taxes.  
d. No transaction costs.   

- 이러한 가정을 기반으로 모든 투자자들이 동일한 efficient frontier를 계산하게 된다. - 각 투자자의 optimal risky portfolio에 기반한 CAL과 CML이 동일  
-> 증권 분석을 건너뛰고 단순히 market portfolio를 보유함으로써 효율적인 포트폴리오를 얻어낼 수 있음  
- 이러한 결과를 mutual fund theorem이라고 한다.   

$y = \frac{E(r_M)-r_f}{A\sigma^2_M}$  
→ CAPM경제 내에서는 순차입과 순대출은 0이 되어야 하기 때문에 위험 포트폴리오의 평균 가중치 $\bar{y} = 1$ 이다.   
$E(R_M) = \bar{A}\sigma^2_M$    

※ Expected Returns on Individual Securities  

시장 포트폴리오의 분산에 대한 GE 주식의 기여도:$w_{\text{GE}} \left[ w_1 \text{Cov}(R_1, R_{\text{GE}}) + w_2 \text{Cov}(R_2, R_{\text{GE}}) + \cdots + w_{\text{GE}} \text{Cov}(R_{\text{GE}}, R_{\text{GE}}) + \cdots + w_n \text{Cov}(R_n, R_{\text{GE}}) \right]
$    

GE에 대한 투자에 대한 위험 대비 보상 비율:  
$\frac{\text{GE's contribution to risk premium}}{\text{GE's contribution to variance}} = \frac{w_{\text{GE}} E(R_{\text{GE}})}{w_{\text{GE}} \text{Cov}(R_{\text{GE}}, R_M)} = \frac{E(R_{\text{GE}})}{\text{Cov}(R_{\text{GE}}, R_M)}
$  

시장 포트폴리오에 대한 투자에 대한 보상 대 위험 비율: $\frac{\text{Market risk premium}}{\text{Market variance}} = \frac{E(R_M)}{\sigma_M^2}$  

- 균형의 기본 원칙은 모든 투자가 동일한 위험 대비 보상 비율을 제공해야 한다는 것  
$\frac{E(R_{\text{GE}})}{\text{Cov}(R_{\text{GE}}, R_M)} = \frac{E(R_M)}{\sigma_M^2}$  
→ $E(R_{\text{GE}}) = \frac{\text{Cov}(R_{\text{GE}}, R_M)}{\sigma_M^2} E(R_M)$  
→ $E(r_{\text{GE}}) = r_f + \beta_{\text{GE}} [E(r_M) - r_f]$  


  
- SML의 기울기: $E(r_M) - r_f$  
- 저평가된 주식은 SML 위에, 고평과된 주식은 SML 밑에 위치   

※ zero-beta portfolio  
$E(r_i) - E(r_Z) = \beta_i[E(r_M) - E(r_Z)]$  
- CAPM에서의 SML의 risk-free rate가 $E(r_Z)$로 대체된 형태  
- $E(r_Z)>r_f$, 즉 SML의 기울기가 더 낮아진다.  


※ Merton's ICAPM expected return-beta equation  

- extra-market risk K를 식별하고 K관련 헤지 포트폴리오를 찾을 수 있다고 가정  
$E(R_i) = \beta_{iM}E(R_M) + \sum^K_{k=1}\beta_{ik}E(R_k)$  

※ A Consumption-Based CAPM  

$E(R_i) = \beta_{iC}RP_C$  
(C: consumption-tracking portfolio, $RP_C$: risk premium associated with consumption uncertainty)  

$RP_C = E(R_C) = E(r_C) - r_f$  

$E(R_M) = \alpha_M + \beta_{MC}E(R_C) + \epsilon_M$  

CCAPM의 장단점  
- 소비 불확실성과 투자 기회의 특성화를 간결하게 통합할 수 있다.  
- 투자 기회의 특성을 특정 매개변수로 나타내므로 투자자들이 이러한 변화를 쉽게 이해하고 분석할 수 있다. 
- 소비 성장 데이터는 상당한 오차로 측정되며 공개 주기도 길다.(최대 월 1회)

- 가중치의 절댓값의 합은 1이 되도록, 가중치의 합은 0이 되도록 설정한다.  

레버리지 비율: 포트폴리오의 레버리지 양을 측정하는 척도로 활용  
→ 롱포지션 금액과 숏포지션 금액을 더한 값을 초기 자본금으로 나눠서 계산  

1. De-mean  
$w'_i = w_i - \frac{1}{n} \sum_{j=1}^{n} w_j$    
포트폴리오가 시장 전체의 움직임에 중립적이 되도록 하기 위함.  
→ 가중치의 합이 0이 된다.  


2. Rescale  
$w''_i = \frac{w'_i}{\sum_{i=1}^{n} \left| w'_i \right|}$    
포트폴리오가 시장의 변동성에 따른 리스크를 최소화하면서도 레버리지를 통해 수익을 극대화할 수 있도록 하기 위함.  
→ 가중치의 절댓값의 합이 1이 된다.  


## Lesson23. Factor Models and Types of Factors  
※ Linear Factor Model  
$r_i = b_{i1}f_1 + b_{i2}f_2 + \cdots + b_{iK}f_K + s_i$  

$\begin{align*}
r_i &= \text{자산 } i \text{의 수익률} \\
f_1 &= \text{팩터 수익률 1의 값} \\
b_{i1} &= \text{팩터 수익률 1의 단위 변화당 자산 } i \text{의 수익률 변화} \\
K &= \text{팩터의 수} \\
s_i &= \text{K 팩터와 관련되지 않은 자산 } i \text{의 수익률 부분}
\end{align*}
$  

※ Assumption  
1. 각 자산의 idiosyncratic 위험 요소들은 서로 상관관계가 없다고 가정   
2. 각 자산의 idiosyncratic는 팩터들로부터 독립적이라고 가정   


1. 알파 팩터 (Alpha Factor)  
자산이나 포트폴리오가 시장이나 벤치마크 대비 초과 수익을 낼 수 있도록 기여하는 요소  


2. 리스크 팩터 (Risk Factor)  
투자자들이 특정 자산이 노출된 위험을 이해하고 관리하기 위해 사용하는 요소  

→ 알파 팩터는 투자자들에게 알려지면서 결국 성과를 잃고 리스크팩터로 변한다.  
알파 팩터가 많이 알려지면 동일한 신호를 발견하고 주식을 매수하기 때문에 기존보다 주가가 상승한 상태에서 매수를 하게 되어 수익률을 기대했던 것 만큼 얻지 못할 수 있다.  
하지만 그만큼 팩터가 알려지게 되면서 시장 가격이 움직이기 때문에 리스크 팩터가 될 수 있다.  


Price-volume  
- Unadjusted or adjusted prices  
- Open, high, low, close  
- Different frequencies 

※ 쇼트 스퀴즈  
주가가 급격히 상승함에 따라, 다른 공매도 투자자들도 손실을 줄이기 위해 서둘러 주식을 매수하게되고 이는 주식 가격을 더 상승시키게 하기 때문에 악순환이 계속 반복되는 현상. 이와 반대의 개념을 **롱 스퀴즈** 라고한다.  


※ Event-Driven Factors  
자연재해, 정부의 주요 변화, 인수합병 등이 해당  
 
※ Alternative Data  
재무제표, 애널리스트 보고서 등을 포함하지 않는 데이터 등 거래 신호를 생성하기 위한 상대적으로 새로운 소스인 데이터   
→ 기업에 대한 대중의 감정을 분석하기 위해 NLP 활용. 뉴스 기사 또는 블로그 등을 통해 해당 주식의 매수 or 매도 분류 가능성이 얼마나 되는지를 알 수 있음.

## Lesson25. Time Series and Cross Sectional Risk Models   

※ 시장 수익률 하나만을 팩터로 사용하는 모델  
- 시장 수익률의 분산을 구하기 위해 시계열 데이터의 분산 계산  
→ 너무 많은 샘플을 사용하면 데이터가 현재 변동성을 추정하지 못할 수 있음. 너무 적은 샘플을 사용하면 데이터에 노이즈 비율이 너무 높아짐.  
 
※ Fama-French 3 Factor Model  

1. Size  
시가총액 기준으로 소규모 회사가 대형 회사에 비해 평균 수익률이 더 높은 경향이 있음. 

시가총액 기준으로 주식을 정렬한 후   
시가총액 하위 10%는 Small, 상위 10%는 Big으로 설정. 각 그룹의 동일 가중 수익률을 계산한 다음 소형주 수익률에서 대형주 수익률을 뺀 값을 2로 나누어준다.  
→ SMB(Small minus Big)  


2. Value  
시장가치 대비 장부가치(B/M)가 높은 주식이 낮은 주식보다 성과가 좋은 경향이 있다.  
해당 가치가 높은 상위 30%의 주식을 Value(High), 하위 30%의 주식을 Growth(Low)라고 한다.  
Factor return = $\frac{r_{value}-r_{growth}}{2}$   
→ HML(High minus Low)  


3. Market  
기존 CAPM과 동일하게 시장 초과수익률 활용   

※ Fama-French 3 Factor Model 과정  
시가총액 기준으로 정렬하여 Small, Big 선정. B/M기준으로 Value, Neutral, Growth으로 정렬하여 6개의 그룹으로 나누어준다.  
1. Small Value  
2. Small Neutral  
3. Small Growth  
4. Big Value  
5. Big Neutral  
6. Big Growth  

SMB = {(1+2+3) - (4+5+6)} / 3  
HML = {(1+4) - (3+6)} / 2  

→ 이후 각 팩터의 공분산 행렬 F 계산  
$f_m =  \quad r_m - r_f$

$f_s = \quad \text{SMB} = \frac{1}{3} \left( (r_{s,v} + r_{s,n} + r_{s,g}) - (r_{b,v} + r_{b,n} + r_{b,g}) \right)$

$f_v = \quad \text{HML} = \frac{1}{2} \left( (r_{s,v} + r_{b,v}) - (r_{s,g} + r_{b,g}) \right)$  


B: n by 3  (n: 자산 개수)  

S: 각 자산의 분산이 대각 원소로 들어간 대각행렬  


※ Cross Section  
한 시점에서의 데이터들을 활용. 

- 시계열 팩터 모델은 팩터 수익률과 자산 수익률을 통해 베타값을 추정하는 반면 횡단면 모델은 베타값과 자산 수익률을 통해 팩터 수익률을 얻어낸다.  


## Lesson 26. Risk Factor Models with PCA  

주성분분석(PCA): 데이터의 분산을 가장 잘 설명하는 서로 독립적인 팩터들을 추출해내는 차원 축소 방법   

각 데이터들을 주성분 축으로 사영시킨 점들간의 분산을 최대화하는것이 목표. 이때 기존 데이터들의 평균을 원점으로 만들기 위해 각 좌표에서 평균값만큼을 빼준다(평균 중심화 or 데이터 정규화).  

- 데이터의 전체 분산은 원점으로부터 기존 데이터간의 거리의 제곱합을 나타낸다.  
- 각 팩터의 방향은 서로 직교한다.  
- 주성분 개수는 팩터의 개수를 넘지 않는다.  

첫번째 주성분의 방향을 찾았다면 해당 방향을 따라 각 데이터 포인트의 구성요소를 찾아 데이터 포인트에서 빼면 첫번째 주성분에 수직인 2D평면에 존재한다.  
→ 구하고자 하는 주성분 개수만큼 해당 과정 반복    

- PCA에서 구한 팩터들의 공분산 행렬은 대각행렬이 된다.   


$X = WY + S$  

X: 기존 데이터 행렬  
W: 각 주성분의 고유벡터로 이루어진 행렬  
Y: 주성분 점수 행렬  
S: 잔차
